### F1 & Accuracy by Template

Loads Claude CSV and computes template-level metrics from `ground_truth_answer` vs `model_answer`.
If the file is missing, this block is skipped.


````python
import pandas as pd
import re

#INPUT = "output_with_answers_recovered.csv"
INPUT = "1_2_with_answers_claude-opus-4_6.csv"
#INPUT = "1_1_1_Eval_Claude_Opus_1.csv"

COL_TEMPLATE = "template_id"
COL_GT = "ground_truth_answer"
COL_PRED = "model_answer"

from pathlib import Path
SKIP_CLAUDE = not Path(INPUT).exists()
if SKIP_CLAUDE:
    print(f"Claude Opus: missing {INPUT}. Cell skipped. Dictionary not saved.")
    df = pd.DataFrame(columns=[COL_TEMPLATE, COL_GT, COL_PRED])
else:
    df = pd.read_csv(INPUT)

def normalize_text(s: str) -> str:
    s = "" if pd.isna(s) else str(s)
    s = s.lower().strip()
    s = re.sub(r"[^\w\s]", " ", s, flags=re.UNICODE)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def normalize_label(s: str) -> str:
    t = normalize_text(s)
    if t in {"yes", "true"}:
        return "yes"
    if t in {"no", "false"}:
        return "no"
    return t

def contains_whole_word(text: str, word: str) -> bool:
    if not word:
        return False
    pattern = r"(?:^|\s)" + re.escape(word) + r"(?:$|\s)"
    return re.search(pattern, text) is not None

df["gt_norm"] = df[COL_GT].apply(normalize_label)
df["pred_norm_text"] = df[COL_PRED].apply(normalize_text)

def predicted_label_from_text(pred_text: str, gt_label: str) -> str:
    return gt_label if contains_whole_word(pred_text, gt_label) else ""

df["pred_label"] = [
    predicted_label_from_text(p, g)
    for p, g in zip(df["pred_norm_text"], df["gt_norm"])
]

# correctness
df["is_correct"] = df["pred_label"] == df["gt_norm"]

# ---- metrics per each template_id ----
def group_metrics(g: pd.DataFrame) -> pd.Series:
    n = len(g)
    acc = g["is_correct"].mean() if n else 0.0

    labels = sorted(set(g["gt_norm"].unique()) | set(g["pred_label"].unique()))
    labels = [lab for lab in labels if lab != ""]  # remove empty label

    # F1:
    # - If only yes/no in GT -> binary F1 (positive=yes)
    # - Otherwise, macro F1 by GT classes (ignoring empty class)
    gt_set = set(g["gt_norm"].unique())
    if gt_set.issubset({"yes", "no"}):
        y_true = g["gt_norm"].tolist()
        y_pred = [("yes" if x == "yes" else "no") if x in {"yes","no"} else "no" for x in g["pred_label"].tolist()]
        # F1 positive=yes
        tp = sum((t == "yes") and (p == "yes") for t, p in zip(y_true, y_pred))
        fp = sum((t != "yes") and (p == "yes") for t, p in zip(y_true, y_pred))
        fn = sum((t == "yes") and (p != "yes") for t, p in zip(y_true, y_pred))
        precision = tp / (tp + fp) if (tp + fp) else 0.0
        recall = tp / (tp + fn) if (tp + fn) else 0.0
        f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0.0
        f1_type = "binary(yes)"
    else:
        # macro F1 per GT classes
        y_true = g["gt_norm"].tolist()
        y_pred = g["pred_label"].tolist()

        f1s = []
        for lab in sorted(gt_set):
            tp = sum((t == lab) and (p == lab) for t, p in zip(y_true, y_pred))
            fp = sum((t != lab) and (p == lab) for t, p in zip(y_true, y_pred))
            fn = sum((t == lab) and (p != lab) for t, p in zip(y_true, y_pred))
            precision = tp / (tp + fp) if (tp + fp) else 0.0
            recall = tp / (tp + fn) if (tp + fn) else 0.0
            f1_lab = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0.0
            f1s.append(f1_lab)
        f1 = sum(f1s) / len(f1s) if f1s else 0.0
        f1_type = "macro(GT-classes)"

    return pd.Series({
        "n": n,
        "accuracy": acc,
        "f1": f1,
        "f1_type": f1_type,
    })

rows = []
if SKIP_CLAUDE:
    report = pd.DataFrame(columns=["n", "accuracy", "f1", "f1_type", COL_TEMPLATE])
else:
    for template_id, g in df.groupby(COL_TEMPLATE, dropna=False):
        m = group_metrics(g)          # Series: n, accuracy, f1, f1_type
        m[COL_TEMPLATE] = template_id
        rows.append(m)
    report = pd.DataFrame(rows).sort_values(COL_TEMPLATE).reset_index(drop=True)
report
````

### JSON Parsing and MVSS Calculation

Defines JSON parsing, key matching, attribute comparison, and MVSS summary scoring.
Each answer is compared by `value`, `unit`, and `source_view` for every matched attribute key.
`strict attribute match` is the exact-match rate at attribute level; `mean MVSS` is the average sample-level partial score, and `MVSS strict accuracy` is the share of samples with fully strict match.


````python
import re
import json
import ast
import numpy as np
import pandas as pd
from decimal import Decimal, InvalidOperation

# =========================
# CONFIG
# =========================

# Set your actual columns
COL_GT = "ground_truth_answer"
COL_PRED = "model_answer"

# Model metadata fields that are not treated as measurements
IGNORE_PRED_KEYS = {
    "reasoning",
    "reasoning_brief",
    "explanation",
    "notes",
    "comment",
}

# If True, extra measurement keys in prediction force sample-level MVSS = 0
PENALIZE_EXTRA_PRED_KEYS = True

# If a GT attribute exists and is null, prediction must explicitly return this attribute as null.
# In other words, a missing key != a correct null value.
REQUIRE_EXPLICIT_NULL = True

# For value comparison after unit conversion
NUMERIC_TOL = Decimal("0.000001")


# =========================
# BASIC NORMALIZATION
# =========================

def _is_null_like(x) -> bool:
    if x is None:
        return True
    if isinstance(x, float) and pd.isna(x):
        return True
    if isinstance(x, str) and x.strip().lower() in {"", "null", "none", "nan"}:
        return True
    return False


def _normalize_text(s: object) -> str:
    if s is None:
        return ""
    s = str(s).strip().lower()
    s = s.replace("_", " ").replace("-", " ")
    s = re.sub(r"[^\w\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s


def _normalize_key_string(key: object) -> str:
    return _normalize_text(key)


def _simple_stem(token: str) -> str:
    if token.endswith("ies") and len(token) > 4:
        return token[:-3] + "y"
    if token.endswith("s") and len(token) > 4:
        return token[:-1]
    return token


STOPWORDS = {
    "the", "a", "an", "of", "to", "from", "for", "in", "on", "at",
    "nearest", "center", "centre", "line", "centerline", "lowest",
    "highest", "left", "right"
}

# Canonical dimension tokens + aliases.
DIMENSION_TOKEN_ALIASES = {
    "width": {"width", "wide", "breadth"},
    "height": {"height", "high", "tall"},
    "depth": {"depth", "deep"},
    "length": {"length", "long"},
    "distance": {"distance", "dist", "gap", "clearance", "spacing"},
    "thickness": {"thickness", "thick"},
}
DIMENSION_TOKENS = set(DIMENSION_TOKEN_ALIASES.keys())
DIMENSION_ALIAS_TO_CANON = {
    alias: canon
    for canon, aliases in DIMENSION_TOKEN_ALIASES.items()
    for alias in aliases
}


def _tokenize_key(key: object) -> list:
    s = _normalize_key_string(key)
    tokens = []
    for tok in s.split():
        tok = _simple_stem(tok)
        if tok and tok not in STOPWORDS:
            tokens.append(tok)
    return tokens


def _extract_dimension_tokens(key: object) -> set:
    dims = set()
    for tok in _tokenize_key(key):
        canon = DIMENSION_ALIAS_TO_CANON.get(tok)
        if canon:
            dims.add(canon)
    return dims


def _primary_dimension_token(key: object):
    dims = sorted(_extract_dimension_tokens(key))
    return dims[0] if len(dims) == 1 else None


def _dimension_token_check(gt_key: object, pred_key: object, pred_key_present: bool) -> dict:
    gt_dim = _primary_dimension_token(gt_key)
    pred_dim = _primary_dimension_token(pred_key) if pred_key_present else None

    # Apply dimension-token correctness only when GT key has a clear single dimension token.
    applicable = gt_dim is not None
    if not applicable:
        match = 1
    elif not pred_key_present:
        match = 0
    else:
        match = int(pred_dim == gt_dim)

    return {
        "gt_dimension_token": gt_dim,
        "pred_dimension_token": pred_dim,
        "dimension_applicable": applicable,
        "dimension_match": match,
    }
````

````python
# =========================
# JSON PARSING / VALIDATION
# =========================

import re
import json
import ast
import unicodedata
import pandas as pd

def _cleanup_json_candidate(s: str, strip_trailing_commas: bool = False) -> str:
    s = unicodedata.normalize("NFKC", s)
    s = s.replace("???", '"').replace("???", '"').replace("??В", "'").replace("???", "'")
    s = re.sub(r"[\u200b-\u200d\ufeff]", "", s)
    s = re.sub(r"^\s*```(?:json)?\s*", "", s, flags=re.IGNORECASE)
    s = re.sub(r"\s*```\s*$", "", s)
    s = s.strip()
    if strip_trailing_commas:
        s = re.sub(r",\s*([}\]])", r"\1", s)
    return s


def _append_missing_closing_braces(s: str) -> str:
    depth = 0
    in_string = False
    escape = False

    for ch in s:
        if in_string:
            if escape:
                escape = False
            elif ch == "\\":
                escape = True
            elif ch == '"':
                in_string = False
        else:
            if ch == '"':
                in_string = True
            elif ch == "{":
                depth += 1
            elif ch == "}":
                if depth > 0:
                    depth -= 1

    if depth > 0:
        s = s + ("}" * depth)
    return s


def _try_parse_json_dict(cand: str):
    if not cand:
        return None
    try:
        obj = json.loads(cand)
        if isinstance(obj, dict):
            return obj
    except Exception:
        pass
    return None


def _try_parse_relaxed_dict(cand: str):
    if not cand:
        return None
    try:
        py_cand = re.sub(r"\bnull\b", "None", cand, flags=re.IGNORECASE)
        py_cand = re.sub(r"\btrue\b", "True", py_cand, flags=re.IGNORECASE)
        py_cand = re.sub(r"\bfalse\b", "False", py_cand, flags=re.IGNORECASE)
        obj = ast.literal_eval(py_cand)
        if isinstance(obj, dict):
            return obj
    except Exception:
        pass
    return None


def _safe_json_load(text: object, allow_repair: bool = False):
    if pd.isna(text):
        return None

    s = _cleanup_json_candidate(str(text), strip_trailing_commas=False)
    if not s:
        return None

    strict_obj = _try_parse_json_dict(s)
    if strict_obj is not None:
        return strict_obj
    if not allow_repair:
        return None

    repaired_candidates = [
        _cleanup_json_candidate(s, strip_trailing_commas=True),
        _append_missing_closing_braces(s),
        _append_missing_closing_braces(_cleanup_json_candidate(s, strip_trailing_commas=True)),
    ]

    seen = set()
    for cand in repaired_candidates:
        if not cand or cand in seen:
            continue
        seen.add(cand)

        obj = _try_parse_json_dict(cand)
        if isinstance(obj, dict):
            return obj

        obj = _try_parse_relaxed_dict(cand)
        if isinstance(obj, dict):
            return obj

    return None


def _unwrap_dimensions(obj):
    if isinstance(obj, dict) and isinstance(obj.get("dimensions"), dict):
        return obj["dimensions"]
    return obj


def _extract_measurement_items(obj: dict, is_prediction: bool = False) -> dict:
    if not isinstance(obj, dict):
        return {}

    obj = _unwrap_dimensions(obj)
    out = {}

    ignore_pred_keys_norm = {_normalize_text(k) for k in IGNORE_PRED_KEYS}

    for k, v in obj.items():
        if is_prediction and _normalize_text(k) in ignore_pred_keys_norm:
            continue
        out[k] = v

    return out


def _is_valid_attr_payload(v) -> bool:
    if _is_null_like(v):
        return True
    if isinstance(v, dict):
        has_known_field = any(k in v for k in ["value", "val", "unit", "source_view", "sourceView", "source"])
        return has_known_field
    return False


def _is_valid_prediction_json(pred_obj) -> bool:
    if not isinstance(pred_obj, dict):
        return False

    pred_items = _extract_measurement_items(pred_obj, is_prediction=True)
    if len(pred_items) == 0:
        return False

    return all(_is_valid_attr_payload(v) for v in pred_items.values())
````

In this notebook, an `attribute` is one measured item in model JSON output for a matched key.

Typical fields:
- `value`
- `unit`
- `source_view`

Used in code paths:
- `_is_valid_attr_payload`
- `_extract_attr_fields`
- `_eval_attribute`


````python
# =========================
# ATTRIBUTE-LEVEL ANALYSIS
# =========================

def _normalize_unit(unit: object) -> str:
    if unit is None:
        return ""
    s = str(unit).strip().lower().replace(" ", "")
    unit_map = {
        "meter": "m",
        "meters": "m",
        "metre": "m",
        "metres": "m",
        "millimeter": "mm",
        "millimeters": "mm",
        "millimetre": "mm",
        "millimetres": "mm",
        "centimeter": "cm",
        "centimeters": "cm",
        "centimetre": "cm",
        "centimetres": "cm",
        "inch": "in",
        "inches": "in",
        "foot": "ft",
        "feet": "ft",
    }
    return unit_map.get(s, s)


UNIT_TO_MM = {
    "mm": Decimal("1"),
    "cm": Decimal("10"),
    "m": Decimal("1000"),
    "in": Decimal("25.4"),
    "ft": Decimal("304.8"),
}


def _normalize_source_view(value: object) -> str:
    return _normalize_text(value)


def _parse_decimal(value: object):
    if value is None:
        return None

    if isinstance(value, (int, float, Decimal)):
        try:
            return Decimal(str(value))
        except (InvalidOperation, ValueError):
            return None

    s = str(value).strip()
    if s == "" or s.lower() in {"null", "none", "nan"}:
        return None

    s = s.replace(",", "")

    try:
        return Decimal(s)
    except (InvalidOperation, ValueError):
        pass

    m = re.search(r"-?\d+(?:\.\d+)?", s)
    if not m:
        return None

    try:
        return Decimal(m.group(0))
    except (InvalidOperation, ValueError):
        return None


def _normalize_scalar(value: object):
    if _is_null_like(value):
        return None

    num = _parse_decimal(value)
    if num is not None:
        norm = format(num.normalize(), "f").rstrip("0").rstrip(".")
        return norm if norm else "0"

    return _normalize_text(value)


def _convert_to_mm(value: object, unit: object):
    num = _parse_decimal(value)
    unit_norm = _normalize_unit(unit)
    if num is None or unit_norm not in UNIT_TO_MM:
        return None
    return num * UNIT_TO_MM[unit_norm]


def _value_match(gt_val, pred_val, gt_unit=None, pred_unit=None) -> int:
    gt_mm = _convert_to_mm(gt_val, gt_unit)
    pred_mm = _convert_to_mm(pred_val, pred_unit)

    # if both can be converted -> compare after conversion
    if gt_mm is not None and pred_mm is not None:
        return int(abs(gt_mm - pred_mm) <= NUMERIC_TOL)

    # fallback to normalized scalar comparison
    return int(_normalize_scalar(gt_val) == _normalize_scalar(pred_val))


def _extract_attr_fields(attr):
    if _is_null_like(attr):
        return True, None, None, None

    if isinstance(attr, dict):
        val = attr.get("value", attr.get("val"))
        unit = attr.get("unit")
        source = attr.get("source_view", attr.get("sourceView", attr.get("source")))

        if _is_null_like(val) and _is_null_like(unit) and _is_null_like(source):
            return True, None, None, None

        return False, val, unit, source

    # if model returned scalar instead of object, treat as value-only payload
    return False, attr, None, None


def _eval_attribute(gt_attr, pred_attr, pred_key_present: bool, key_match: bool = True):
    gt_null, gt_val, gt_unit, gt_src = _extract_attr_fields(gt_attr)

    value_applicable = not gt_null
    unit_applicable = value_applicable and (not _is_null_like(gt_unit))
    source_applicable = value_applicable and (not _is_null_like(gt_src))

    if not pred_key_present:
        if gt_null and not REQUIRE_EXPLICIT_NULL and key_match:
            null_match = 1
            strict_match = 1
        else:
            null_match = 0
            strict_match = 0

        return {
            "null_match": null_match,
            "value_match": 0,
            "unit_match": 0 if unit_applicable else 0,
            "source_match": 0 if source_applicable else 0,
            "strict_match": strict_match,
            "value_applicable": value_applicable,
            "unit_applicable": unit_applicable,
            "source_applicable": source_applicable,
            "gt_is_null": gt_null,
            "pred_is_null": True,
            "missing_pred_key": True,
        }

    pred_null, pred_val, pred_unit, pred_src = _extract_attr_fields(pred_attr)

    # If dimension token mismatched, this attribute is automatically incorrect.
    if not key_match:
        if gt_null:
            return {
                "null_match": 0,
                "value_match": 0,
                "unit_match": 0,
                "source_match": 0,
                "strict_match": 0,
                "value_applicable": False,
                "unit_applicable": False,
                "source_applicable": False,
                "gt_is_null": True,
                "pred_is_null": pred_null,
                "missing_pred_key": False,
            }

        return {
            "null_match": 0,
            "value_match": 0,
            "unit_match": 0 if unit_applicable else 0,
            "source_match": 0 if source_applicable else 0,
            "strict_match": 0,
            "value_applicable": value_applicable,
            "unit_applicable": unit_applicable,
            "source_applicable": source_applicable,
            "gt_is_null": False,
            "pred_is_null": pred_null,
            "missing_pred_key": False,
        }

    if gt_null:
        null_match = int(pred_null)
        strict_match = int(pred_null)
        return {
            "null_match": null_match,
            "value_match": null_match,
            "unit_match": null_match,
            "source_match": null_match,
            "strict_match": strict_match,
            "value_applicable": False,
            "unit_applicable": False,
            "source_applicable": False,
            "gt_is_null": True,
            "pred_is_null": pred_null,
            "missing_pred_key": False,
        }

    # gt is not null
    value_match = _value_match(gt_val, pred_val, gt_unit, pred_unit)

    if unit_applicable:
        unit_match = int(_normalize_unit(gt_unit) == _normalize_unit(pred_unit))
    else:
        unit_match = 1

    if source_applicable:
        source_match = int(_normalize_source_view(gt_src) == _normalize_source_view(pred_src))
    else:
        source_match = 1

    strict_match = int(value_match and unit_match and source_match)

    return {
        "null_match": int(gt_null == pred_null),
        "value_match": value_match,
        "unit_match": unit_match,
        "source_match": source_match,
        "strict_match": strict_match,
        "value_applicable": value_applicable,
        "unit_applicable": unit_applicable,
        "source_applicable": source_applicable,
        "gt_is_null": False,
        "pred_is_null": pred_null,
        "missing_pred_key": False,
    }
````

A `key` is the measurement/property name in JSON (for example, `table_height_cm`).

Evaluation flow:
1. Match GT keys to predicted keys (`_match_gt_to_pred_keys`).
2. Compare matched attributes (`gt_attr` vs `pred_attr`).

Example:
```json
{
  "dimensions": {
    "table_height_cm": {"value": 74, "unit": "cm", "source_view": "front"},
    "distance_chair_to_table_cm": {"value": 30, "unit": "cm", "source_view": "top"}
  }
}
```


````python
# =========================
# KEY MATCHING
# =========================

def _key_match_score(gt_key: str, pred_key: str) -> int:
    gt_norm = _normalize_key_string(gt_key)
    pred_norm = _normalize_key_string(pred_key)

    if gt_norm == pred_norm:
        return 1000

    gt_tokens = set(_tokenize_key(gt_key))
    pred_tokens = set(_tokenize_key(pred_key))

    common = gt_tokens & pred_tokens
    if not common:
        return 0

    gt_dims = _extract_dimension_tokens(gt_key)
    pred_dims = _extract_dimension_tokens(pred_key)

    # If GT has explicit dimension token, prediction must also specify one.
    if gt_dims and not pred_dims:
        return 0

    # If both explicitly specify dimensions, they must overlap.
    if gt_dims and pred_dims and not (gt_dims & pred_dims):
        return 0

    subset_match = gt_tokens.issubset(pred_tokens) or pred_tokens.issubset(gt_tokens)

    # avoid matching on one weak/common token unless one key is subset of another
    if len(common) < 2 and not subset_match and not (gt_dims & pred_dims):
        return 0

    score = 0
    score += 25 * len(gt_dims & pred_dims)
    score += 5 * len(common)
    if subset_match:
        score += 10
    if gt_dims and pred_dims:
        score += 10

    return score


def _match_gt_to_pred_keys(gt_keys, pred_keys):
    candidates = []
    for gk in gt_keys:
        for pk in pred_keys:
            score = _key_match_score(gk, pk)
            if score > 0:
                candidates.append((score, gk, pk))

    # greedy matching by best score first
    candidates = sorted(candidates, key=lambda x: (-x[0], x[1], x[2]))

    gt_to_pred = {}
    used_gt = set()
    used_pred = set()

    for score, gk, pk in candidates:
        if gk in used_gt or pk in used_pred:
            continue
        gt_to_pred[gk] = pk
        used_gt.add(gk)
        used_pred.add(pk)

    unmatched_gt = [gk for gk in gt_keys if gk not in gt_to_pred]
    unmatched_pred = [pk for pk in pred_keys if pk not in used_pred]

    return gt_to_pred, unmatched_gt, unmatched_pred
````

````python
# =========================
# SOFT FALLBACK FOR UNMATCHED KEYS
# =========================

ENABLE_SOFT_ATTR_FALLBACK = True
SOFT_ATTR_FALLBACK_MIN_SCORE = 30
SOFT_MATCH_WEIGHT = 0.5   # half-credit weight

def _soft_attr_match_score(gt_key, gt_attr, pred_key, pred_attr) -> int:
    gt_null, gt_val, gt_unit, gt_src = _extract_attr_fields(gt_attr)
    pred_null, pred_val, pred_unit, pred_src = _extract_attr_fields(pred_attr)

    if gt_null != pred_null:
        return 0

    if gt_null and pred_null:
        return 20

    score = 0

    if _value_match(gt_val, pred_val, gt_unit, pred_unit):
        score += 20

    if (not _is_null_like(gt_unit)) and (not _is_null_like(pred_unit)):
        if _normalize_unit(gt_unit) == _normalize_unit(pred_unit):
            score += 5

    if (not _is_null_like(gt_src)) and (not _is_null_like(pred_src)):
        if _normalize_source_view(gt_src) == _normalize_source_view(pred_src):
            score += 10

    return score


def _soft_match_unmatched_by_attribute(gt_items, pred_items, unmatched_gt, unmatched_pred, min_score=30):
    candidates = []

    for gk in unmatched_gt:
        for pk in unmatched_pred:
            score = _soft_attr_match_score(
                gt_key=gk,
                gt_attr=gt_items.get(gk),
                pred_key=pk,
                pred_attr=pred_items.get(pk),
            )
            if score >= min_score:
                candidates.append((score, gk, pk))

    candidates = sorted(candidates, key=lambda x: (-x[0], x[1], x[2]))

    soft_matches = {}
    used_gt = set()
    used_pred = set()

    for score, gk, pk in candidates:
        if gk in used_gt or pk in used_pred:
            continue
        soft_matches[gk] = pk
        used_gt.add(gk)
        used_pred.add(pk)

    new_unmatched_gt = [gk for gk in unmatched_gt if gk not in soft_matches]
    new_unmatched_pred = [pk for pk in unmatched_pred if pk not in used_pred]

    return soft_matches, new_unmatched_gt, new_unmatched_pred
````

### Claude Evaluation

Builds final Claude metric dictionaries used later in cross-model summary assembly.


````python
# =========================
# MAIN EVALUATION
# =========================

def evaluate_mvss_dataframe(df_in: pd.DataFrame, col_gt: str, col_pred: str):
    n = len(df_in)

    component_totals = {
        "total_value_count": 0,
        "correct_value_count": 0,
        "total_unit_count": 0,
        "correct_unit_count": 0,
        "total_source_view_count": 0,
        "correct_source_view_count": 0,
        "total_null_count": 0,
        "correct_null_count": 0,
        "total_dimension_token_count": 0,
        "correct_dimension_token_count": 0,
        "total_attribute_count": 0,
        "strict_attribute_match_sum": 0,
    }

    sample_rows = []
    attr_rows = []

    question_id_candidates = [
        "generated_question_id",
        "question_number",
        "question_id",
    ]
    question_number_col = next((c for c in question_id_candidates if c in df_in.columns), None)

    for idx, row in df_in.iterrows():
        question_number = row.get(question_number_col) if question_number_col else idx
        if pd.isna(question_number):
            question_number = idx

        gt_obj = _safe_json_load(row.get(col_gt), allow_repair=False)

        strict_pred_obj = _safe_json_load(row.get(col_pred), allow_repair=False)
        repaired_pred_obj = strict_pred_obj if strict_pred_obj is not None else _safe_json_load(row.get(col_pred), allow_repair=True)
        pred_obj = repaired_pred_obj

        gt_items = _extract_measurement_items(gt_obj, is_prediction=False)
        pred_items = _extract_measurement_items(pred_obj, is_prediction=True)

        gt_keys = list(gt_items.keys())
        pred_keys = list(pred_items.keys())

        strict_gt_to_pred, unmatched_gt_before_soft, unmatched_pred_before_soft = _match_gt_to_pred_keys(gt_keys, pred_keys)
        gt_to_pred = dict(strict_gt_to_pred)

        soft_matches = {}
        unmatched_gt_after_soft = list(unmatched_gt_before_soft)
        unmatched_pred_after_soft = list(unmatched_pred_before_soft)

        if ENABLE_SOFT_ATTR_FALLBACK and unmatched_gt_after_soft and unmatched_pred_after_soft:
            soft_matches, unmatched_gt_after_soft, unmatched_pred_after_soft = _soft_match_unmatched_by_attribute(
                gt_items=gt_items,
                pred_items=pred_items,
                unmatched_gt=unmatched_gt_after_soft,
                unmatched_pred=unmatched_pred_after_soft,
                min_score=SOFT_ATTR_FALLBACK_MIN_SCORE,
            )
            gt_to_pred.update(soft_matches)

        sample_json_valid = int(_is_valid_prediction_json(strict_pred_obj))
        pred_was_repaired = int(strict_pred_obj is None and repaired_pred_obj is not None)

        strict_list = []
        partial_credit_list = []

        for gt_key in gt_keys:
            pred_key = gt_to_pred.get(gt_key, None)
            pred_key_present = pred_key is not None
            pred_attr = pred_items.get(pred_key) if pred_key_present else None

            match_mode = "unmatched"
            if gt_key in soft_matches:
                match_mode = "soft_attr"
            elif gt_key in strict_gt_to_pred:
                match_mode = "key_match"

            if match_mode == "soft_attr":
                dim_check = {
                    "gt_dimension_token": _primary_dimension_token(gt_key),
                    "pred_dimension_token": _primary_dimension_token(pred_key),
                    "dimension_applicable": False,
                    "dimension_match": 1,
                }
                key_match_for_eval = True
            else:
                dim_check = _dimension_token_check(
                    gt_key=gt_key,
                    pred_key=pred_key,
                    pred_key_present=pred_key_present,
                )
                key_match_for_eval = bool(dim_check["dimension_match"])

            res = _eval_attribute(
                gt_attr=gt_items.get(gt_key),
                pred_attr=pred_attr,
                pred_key_present=pred_key_present,
                key_match=key_match_for_eval,
            )

            strict_match_base = int(res["strict_match"])
            strict_match = 0 if match_mode == "soft_attr" else strict_match_base
            partial_credit = (float(strict_match_base) * SOFT_MATCH_WEIGHT) if match_mode == "soft_attr" else float(strict_match)

            component_totals["total_attribute_count"] += 1
            component_totals["strict_attribute_match_sum"] += strict_match

            if dim_check["dimension_applicable"]:
                component_totals["total_dimension_token_count"] += 1
                component_totals["correct_dimension_token_count"] += dim_check["dimension_match"]

            if res["gt_is_null"]:
                component_totals["total_null_count"] += 1
                component_totals["correct_null_count"] += res["null_match"]
            else:
                component_totals["total_value_count"] += 1
                component_totals["correct_value_count"] += res["value_match"]

                if res["unit_applicable"]:
                    component_totals["total_unit_count"] += 1
                    component_totals["correct_unit_count"] += res["unit_match"]

                if res["source_applicable"]:
                    component_totals["total_source_view_count"] += 1
                    component_totals["correct_source_view_count"] += res["source_match"]

            strict_list.append(strict_match)
            partial_credit_list.append(partial_credit)

            attr_rows.append({
                "row_index": idx,
                "question_number": question_number,
                "pred_was_repaired": pred_was_repaired,
                "gt_key": gt_key,
                "pred_key": pred_key,
                "matched_key": int(pred_key_present),
                "gt_attr": gt_items.get(gt_key),
                "pred_attr": pred_attr,
                "gt_dimension_token": dim_check["gt_dimension_token"],
                "pred_dimension_token": dim_check["pred_dimension_token"],
                "dimension_applicable": dim_check["dimension_applicable"],
                "dimension_match": dim_check["dimension_match"],
                "missing_pred_key": res["missing_pred_key"],
                "gt_is_null": res["gt_is_null"],
                "pred_is_null": res["pred_is_null"],
                "null_match": res["null_match"],
                "value_match": res["value_match"],
                "unit_match": res["unit_match"],
                "source_match": res["source_match"],
                "strict_match": strict_match,
                "strict_match_base": strict_match_base,
                "match_mode": match_mode,
                "was_soft_matched": int(match_mode == "soft_attr"),
                "partial_credit": partial_credit,
            })

        partial_eval_list = list(partial_credit_list)
        if PENALIZE_EXTRA_PRED_KEYS and len(unmatched_pred_after_soft) > 0:
            partial_eval_list.extend([0.0] * len(unmatched_pred_after_soft))

        sample_partial = float(np.mean(partial_eval_list)) if partial_eval_list else 0.0
        sample_mvss = int(all(v == 1 for v in strict_list)) if strict_list else 0
        if PENALIZE_EXTRA_PRED_KEYS and len(unmatched_pred_after_soft) > 0:
            sample_mvss = 0

        sample_rows.append({
            "row_index": idx,
            "question_number": question_number,
            "json_valid": sample_json_valid,
            "pred_was_repaired": pred_was_repaired,
            "gt_key_count": len(gt_keys),
            "pred_key_count": len(pred_keys),
            "matched_key_count_before_soft": len(strict_gt_to_pred),
            "matched_key_count_after_soft": len(gt_to_pred),
            "matched_key_count": len(gt_to_pred),
            "unmatched_gt_keys_before_soft": unmatched_gt_before_soft,
            "unmatched_pred_keys_before_soft": unmatched_pred_before_soft,
            "unmatched_gt_keys_after_soft": unmatched_gt_after_soft,
            "unmatched_pred_keys_after_soft": unmatched_pred_after_soft,
            "unmatched_gt_keys": unmatched_gt_after_soft,
            "unmatched_pred_keys": unmatched_pred_after_soft,
            "extra_pred_key_count": len(unmatched_pred_after_soft),
            "sample_partial": sample_partial,
            "sample_mvss": sample_mvss,
            "soft_match_count": len(soft_matches),
        })

    total_value_count = component_totals["total_value_count"]
    total_unit_count = component_totals["total_unit_count"]
    total_source_count = component_totals["total_source_view_count"]
    total_null_count = component_totals["total_null_count"]
    total_dim_count = component_totals["total_dimension_token_count"]
    total_attr_count = component_totals["total_attribute_count"]

    component_stats = {
        "n": n,
        "total_value_count": int(total_value_count),
        "correct_value_count": int(component_totals["correct_value_count"]),
        "value_accuracy": (component_totals["correct_value_count"] / total_value_count) if total_value_count else 0.0,
        "total_unit_count": int(total_unit_count),
        "correct_unit_count": int(component_totals["correct_unit_count"]),
        "unit_accuracy": (component_totals["correct_unit_count"] / total_unit_count) if total_unit_count else 0.0,
        "total_source_view_count": int(total_source_count),
        "correct_source_view_count": int(component_totals["correct_source_view_count"]),
        "source_view_accuracy": (component_totals["correct_source_view_count"] / total_source_count) if total_source_count else 0.0,
        "total_null_count": int(total_null_count),
        "correct_null_count": int(component_totals["correct_null_count"]),
        "null_accuracy": (component_totals["correct_null_count"] / total_null_count) if total_null_count else 0.0,
        "total_dimension_token_count": int(total_dim_count),
        "correct_dimension_token_count": int(component_totals["correct_dimension_token_count"]),
        "dimension_token_accuracy": (component_totals["correct_dimension_token_count"] / total_dim_count) if total_dim_count else 0.0,
        "total_attribute_count": int(total_attr_count),
        "strict_attribute_match_accuracy": (component_totals["strict_attribute_match_sum"] / total_attr_count) if total_attr_count else 0.0,
    }

    sample_df = pd.DataFrame(sample_rows)
    attr_df = pd.DataFrame(attr_rows)

    mvss_summary = {
        "n": n,
        "JSON Validity": float(sample_df["json_valid"].mean()) if len(sample_df) else 0.0,
        "Dimension Token Accuracy": component_stats["dimension_token_accuracy"],
        "Strict Attribute Match": component_stats["strict_attribute_match_accuracy"],
        "MVSS_accuracy": float(sample_df["sample_mvss"].mean()) if len(sample_df) else 0.0,
        "MVSS_partial": float(sample_df["sample_partial"].mean()) if len(sample_df) else 0.0,
        "mvss_1_count": int((sample_df["sample_mvss"] == 1).sum()) if len(sample_df) else 0,
        "mvss_0_count": int((sample_df["sample_mvss"] == 0).sum()) if len(sample_df) else 0,
        "repaired_pred_count": int(sample_df["pred_was_repaired"].sum()) if "pred_was_repaired" in sample_df.columns else 0,
        "soft_matched_key_count": int(sample_df["soft_match_count"].sum()) if "soft_match_count" in sample_df.columns else 0,
        "unmatched_gt_key_count_before_soft": int(sample_df["unmatched_gt_keys_before_soft"].apply(lambda x: len(x) if isinstance(x, list) else 0).sum()) if "unmatched_gt_keys_before_soft" in sample_df.columns else 0,
        "unmatched_gt_key_count_after_soft": int(sample_df["unmatched_gt_keys_after_soft"].apply(lambda x: len(x) if isinstance(x, list) else 0).sum()) if "unmatched_gt_keys_after_soft" in sample_df.columns else 0,
        "unmatched_pred_key_count_before_soft": int(sample_df["unmatched_pred_keys_before_soft"].apply(lambda x: len(x) if isinstance(x, list) else 0).sum()) if "unmatched_pred_keys_before_soft" in sample_df.columns else 0,
        "unmatched_pred_key_count_after_soft": int(sample_df["unmatched_pred_keys_after_soft"].apply(lambda x: len(x) if isinstance(x, list) else 0).sum()) if "unmatched_pred_keys_after_soft" in sample_df.columns else 0,
    }

    return component_stats, mvss_summary, sample_df, attr_df
````

````python
import json
import pandas as pd
from pathlib import Path

# ========= CONFIG =========
OUTPUT_SAMPLE_CSV = "1_2_multi_view_dimensions_question_audit.csv"
OUTPUT_ATTR_CSV = "1_2_multi_view_dimensions_attribute_audit.csv"
OUTPUT_AUDIT_XLSX = "1_2_multi_view_dimensions_audit.xlsx"

SCENE_ID_CANDIDATES = [
    "scene_id", "SceneID", "sceneId", "scene", "scene_ID",
    "image_id", "ImageID", "sample_id", "SampleID"
]
QUESTION_COL_CANDIDATES = ["question_text", "question", "prompt", "questionText"]

# ========= CHECK =========
required_vars = ["df_json", "COL_GT", "COL_PRED"]
missing_vars = [v for v in required_vars if v not in globals()]
if missing_vars:
    raise RuntimeError(
        f"Run the Multi-view dimensions evaluation cell first. Missing variables: {missing_vars}"
    )

if len(df_json) == 0:
    raise RuntimeError("df_json is empty: no multi_view_dimensions rows found.")

def _pick_existing_column(df_, candidates, fallback=None):
    for c in candidates:
        if c in df_.columns:
            return c
    return fallback

def _safe_ratio(num, den):
    return float(num / den) if den else 0.0

def _as_list(v):
    if isinstance(v, list):
        return v
    if isinstance(v, tuple):
        return list(v)
    if isinstance(v, set):
        return sorted(v)
    return []

def _to_jsonish(v):
    if isinstance(v, (list, tuple, dict)):
        return json.dumps(v, ensure_ascii=False)
    return v

def _fmt_blob_value(v):
    if isinstance(v, float):
        s = f"{v:.6f}".rstrip("0").rstrip(".")
        return s if s else "0"
    if isinstance(v, (list, tuple, dict)):
        return json.dumps(v, ensure_ascii=False)
    if pd.isna(v):
        return ""
    return str(v)

def _build_blob(metric_dict: dict) -> str:
    order = [
        "row_index",
        "question_number",
        "scene_id",
        "dimension_accuracy",
        "json_validity",
        "strict_attribute_match",
        "MVSS_accuracy",
        "MVSS_partial",
        "repaired_pred_count",
        "soft_matched_key_count",
        "unmatched_gt_key_count_before_soft",
        "unmatched_gt_key_count_after_soft",
        "unmatched_pred_key_count_before_soft",
        "unmatched_pred_key_count_after_soft",
        "total_value_count",
        "correct_value_count",
        "value_accuracy",
        "total_unit_count",
        "correct_unit_count",
        "unit_accuracy",
        "total_source_view_count",
        "correct_source_view_count",
        "source_view_accuracy",
        "total_null_count",
        "correct_null_count",
        "null_accuracy",
        "total_attribute_count",
        "strict_attribute_match_accuracy",
        "gt_key_count",
        "pred_key_count",
        "extra_pred_key_count",
        "unmatched_gt_keys",
        "unmatched_pred_keys",
    ]
    return "\n".join(f"{k}: {_fmt_blob_value(metric_dict.get(k, ''))}" for k in order)

def _row_component_stats(attr_sub: pd.DataFrame) -> dict:
    total_value_count = 0
    correct_value_count = 0
    total_unit_count = 0
    correct_unit_count = 0
    total_source_view_count = 0
    correct_source_view_count = 0
    total_null_count = 0
    correct_null_count = 0
    total_dimension_token_count = 0
    correct_dimension_token_count = 0
    total_attribute_count = 0
    strict_attribute_match_sum = 0

    for _, arow in attr_sub.iterrows():
        total_attribute_count += 1
        strict_attribute_match_sum += int(arow.get("strict_match", 0) or 0)

        if bool(arow.get("dimension_applicable", False)):
            total_dimension_token_count += 1
            correct_dimension_token_count += int(arow.get("dimension_match", 0) or 0)

        gt_null, _, gt_unit, gt_src = _extract_attr_fields(arow.get("gt_attr"))
        if gt_null:
            total_null_count += 1
            correct_null_count += int(arow.get("null_match", 0) or 0)
        else:
            total_value_count += 1
            correct_value_count += int(arow.get("value_match", 0) or 0)

            if not _is_null_like(gt_unit):
                total_unit_count += 1
                correct_unit_count += int(arow.get("unit_match", 0) or 0)

            if not _is_null_like(gt_src):
                total_source_view_count += 1
                correct_source_view_count += int(arow.get("source_match", 0) or 0)

    strict_acc = _safe_ratio(strict_attribute_match_sum, total_attribute_count)
    return {
        "dimension_accuracy": _safe_ratio(correct_dimension_token_count, total_dimension_token_count),
        "total_value_count": int(total_value_count),
        "correct_value_count": int(correct_value_count),
        "value_accuracy": _safe_ratio(correct_value_count, total_value_count),
        "total_unit_count": int(total_unit_count),
        "correct_unit_count": int(correct_unit_count),
        "unit_accuracy": _safe_ratio(correct_unit_count, total_unit_count),
        "total_source_view_count": int(total_source_view_count),
        "correct_source_view_count": int(correct_source_view_count),
        "source_view_accuracy": _safe_ratio(correct_source_view_count, total_source_view_count),
        "total_null_count": int(total_null_count),
        "correct_null_count": int(correct_null_count),
        "null_accuracy": _safe_ratio(correct_null_count, total_null_count),
        "total_attribute_count": int(total_attribute_count),
        "strict_attribute_match": strict_acc,
        "strict_attribute_match_accuracy": strict_acc,
    }

def _first_nonempty(*vals):
    for v in vals:
        if pd.isna(v):
            continue
        s = str(v).strip()
        if s:
            return v
    return ""

# Always recompute through evaluate_mvss_dataframe to keep logic aligned.
_component_stats, _mvss_summary, _sample_df, _attr_df = evaluate_mvss_dataframe(df_json, COL_GT, COL_PRED)

scene_col = _pick_existing_column(df_json, SCENE_ID_CANDIDATES)
question_col = _pick_existing_column(df_json, QUESTION_COL_CANDIDATES)
model_col = "model" if "model" in df_json.columns else None

default_model = _first_nonempty(
    MODEL_NAME if "MODEL_NAME" in globals() else "",
    df_json[model_col].dropna().astype(str).str.strip().iloc[0] if model_col and df_json[model_col].notna().any() else "",
    "unknown_model",
)

base_cols = []
for c in [scene_col, question_col, model_col, COL_GT, COL_PRED]:
    if c and c not in base_cols:
        base_cols.append(c)
base_info = df_json[base_cols].copy() if base_cols else pd.DataFrame(index=df_json.index)
rename_map = {}
if scene_col:
    rename_map[scene_col] = "scene_id"
if question_col:
    rename_map[question_col] = "question_text"
if model_col:
    rename_map[model_col] = "model"
if COL_GT in base_info.columns:
    rename_map[COL_GT] = "ground_truth_answer"
if COL_PRED in base_info.columns:
    rename_map[COL_PRED] = "model_answer"
base_info = base_info.rename(columns=rename_map)

sample_export = _sample_df.copy().merge(base_info, left_on="row_index", right_index=True, how="left")
if "model" not in sample_export.columns:
    sample_export["model"] = default_model
sample_export["model"] = sample_export["model"].fillna("").astype(str).str.strip()
sample_export.loc[sample_export["model"] == "", "model"] = str(default_model)

# Build compact blob per row for manual visual inspection.
question_rows = []
for _, srow in sample_export.iterrows():
    row_index = srow.get("row_index")
    attr_sub = _attr_df[_attr_df["row_index"] == row_index].copy()
    row_stats = _row_component_stats(attr_sub)

    unmatched_gt_before = _as_list(srow.get("unmatched_gt_keys_before_soft"))
    unmatched_pred_before = _as_list(srow.get("unmatched_pred_keys_before_soft"))
    unmatched_gt_after = _as_list(srow.get("unmatched_gt_keys_after_soft"))
    unmatched_pred_after = _as_list(srow.get("unmatched_pred_keys_after_soft"))

    metric_dict = {
        "row_index": row_index,
        "question_number": srow.get("question_number", row_index),
        "scene_id": srow.get("scene_id", ""),
        "json_validity": int(srow.get("json_valid", 0) or 0),
        "MVSS_accuracy": float(srow.get("sample_mvss", 0.0) or 0.0),
        "MVSS_partial": float(srow.get("sample_partial", 0.0) or 0.0),
        "repaired_pred_count": int(srow.get("pred_was_repaired", 0) or 0),
        "soft_matched_key_count": int(srow.get("soft_match_count", 0) or 0),
        "unmatched_gt_key_count_before_soft": len(unmatched_gt_before),
        "unmatched_gt_key_count_after_soft": len(unmatched_gt_after),
        "unmatched_pred_key_count_before_soft": len(unmatched_pred_before),
        "unmatched_pred_key_count_after_soft": len(unmatched_pred_after),
        "gt_key_count": int(srow.get("gt_key_count", 0) or 0),
        "pred_key_count": int(srow.get("pred_key_count", 0) or 0),
        "extra_pred_key_count": int(srow.get("extra_pred_key_count", 0) or 0),
        "unmatched_gt_keys": unmatched_gt_after,
        "unmatched_pred_keys": unmatched_pred_after,
    }
    metric_dict.update(row_stats)

    question_rows.append({
        "model": srow.get("model", default_model),
        "row_index": row_index,
        "question_number": metric_dict["question_number"],
        "scene_id": metric_dict["scene_id"],
        "question_text": srow.get("question_text", ""),
        "ground_truth_answer": srow.get("ground_truth_answer", ""),
        "model_answer": srow.get("model_answer", ""),
        "json_validity": metric_dict["json_validity"],
        "MVSS_accuracy": metric_dict["MVSS_accuracy"],
        "MVSS_partial": metric_dict["MVSS_partial"],
        "repaired_pred_count": metric_dict["repaired_pred_count"],
        "soft_matched_key_count": metric_dict["soft_matched_key_count"],
        "unmatched_gt_keys": metric_dict["unmatched_gt_keys"],
        "unmatched_pred_keys": metric_dict["unmatched_pred_keys"],
        "metrics_blob": _build_blob(metric_dict),
    })

question_export = pd.DataFrame(question_rows)
q_order = [
    "model",
    "row_index",
    "question_number",
    "scene_id",
    "question_text",
    "ground_truth_answer",
    "model_answer",
    "json_validity",
    "MVSS_accuracy",
    "MVSS_partial",
    "repaired_pred_count",
    "soft_matched_key_count",
    "unmatched_gt_keys",
    "unmatched_pred_keys",
    "metrics_blob",
]
q_order = [c for c in q_order if c in question_export.columns]
question_export = question_export[q_order].copy()

# Attribute-level export with model and optional raw context.
attr_export = _attr_df.copy().merge(
    sample_export[["row_index", "model", "ground_truth_answer", "model_answer"]].drop_duplicates(),
    on="row_index",
    how="left",
)
if "model" not in attr_export.columns:
    attr_export["model"] = default_model
attr_export["model"] = attr_export["model"].fillna("").astype(str).str.strip()
attr_export.loc[attr_export["model"] == "", "model"] = str(default_model)

attr_cols = [
    "model",
    "row_index",
    "question_number",
    "pred_was_repaired",
    "gt_key",
    "pred_key",
    "match_mode",
    "was_soft_matched",
    "partial_credit",
    "strict_match",
    "strict_match_base",
    "dimension_match",
    "value_match",
    "unit_match",
    "source_match",
    "null_match",
    "gt_attr",
    "pred_attr",
    "ground_truth_answer",
    "model_answer",
]
attr_cols = [c for c in attr_cols if c in attr_export.columns]
attr_export = attr_export[attr_cols].copy()

for c in question_export.columns:
    if c not in {"ground_truth_answer", "model_answer", "question_text", "metrics_blob", "model"}:
        question_export[c] = question_export[c].apply(_to_jsonish)
for c in attr_export.columns:
    if c not in {"ground_truth_answer", "model_answer", "model", "gt_key", "pred_key", "match_mode"}:
        attr_export[c] = attr_export[c].apply(_to_jsonish)

question_export.to_csv(OUTPUT_SAMPLE_CSV, index=False, encoding="utf-8-sig")
attr_export.to_csv(OUTPUT_ATTR_CSV, index=False, encoding="utf-8-sig")

with pd.ExcelWriter(OUTPUT_AUDIT_XLSX, engine="openpyxl") as writer:
    question_export.to_excel(writer, sheet_name="Multi-view dimension question audit", index=False)
    attr_export.to_excel(writer, sheet_name="Multi-view dimension audit", index=False)

print(f"Saved: {Path(OUTPUT_SAMPLE_CSV).resolve()}")
print(f"Saved: {Path(OUTPUT_ATTR_CSV).resolve()}")
print(f"Saved: {Path(OUTPUT_AUDIT_XLSX).resolve()}")
display(question_export.head(10))
display(attr_export.head(20))
````

### Save Dictionary

````python
import pandas as pd

if "_compute_metrics" not in globals():
    def _compute_metrics(sub: pd.DataFrame) -> dict:
        return {"n": int(len(sub))}

if "_group_sections" not in globals():
    def _group_sections(df_in: pd.DataFrame, col: str, prefix: str) -> dict:
        if col not in df_in.columns:
            return {}
        vals = df_in[col].fillna("").astype(str)
        sections = {}
        for v in sorted([v for v in vals.unique() if v != ""]):
            sections[f"{prefix}={v}"] = _compute_metrics(df_in[vals == v])
        return sections

def _collect_unique_keys(series: pd.Series) -> list:
    out = []
    seen = set()
    for x in series.tolist():
        if isinstance(x, (list, tuple, set)):
            vals = x
        elif pd.isna(x):
            vals = []
        else:
            vals = [x]
        for v in vals:
            s = str(v).strip()
            if s and s not in seen:
                seen.add(s)
                out.append(s)
    return out

def _build_mvss_overall_metrics(component_stats: dict, mvss_summary: dict, sample_df: pd.DataFrame) -> dict:
    sample_df = sample_df if isinstance(sample_df, pd.DataFrame) else pd.DataFrame()

    def _as_list(v):
        if isinstance(v, list):
            return v
        if isinstance(v, tuple):
            return list(v)
        if isinstance(v, set):
            return sorted(v)
        return []

    def _sum_list_lengths(col: str) -> int:
        if col not in sample_df.columns:
            return 0
        return int(sample_df[col].apply(lambda x: len(_as_list(x))).sum())

    def _collect_question_numbers(mask: pd.Series) -> list:
        if "question_number" not in sample_df.columns:
            return []
        out = []
        seen = set()
        for _, r in sample_df.loc[mask].iterrows():
            q = r.get("question_number")
            if pd.isna(q):
                continue
            if isinstance(q, float) and q.is_integer():
                q = int(q)
            qs = str(q).strip()
            if qs and qs not in seen:
                seen.add(qs)
                out.append(qs)
        return out

    unmatched_gt_before_col = "unmatched_gt_keys_before_soft"
    unmatched_pred_before_col = "unmatched_pred_keys_before_soft"
    unmatched_gt_after_col = "unmatched_gt_keys_after_soft" if "unmatched_gt_keys_after_soft" in sample_df.columns else "unmatched_gt_keys"
    unmatched_pred_after_col = "unmatched_pred_keys_after_soft" if "unmatched_pred_keys_after_soft" in sample_df.columns else "unmatched_pred_keys"

    unmatched_gt_after_keys = _collect_unique_keys(sample_df[unmatched_gt_after_col]) if unmatched_gt_after_col in sample_df.columns else []
    unmatched_pred_after_keys = _collect_unique_keys(sample_df[unmatched_pred_after_col]) if unmatched_pred_after_col in sample_df.columns else []

    unmatched_gt_before_keys = _collect_unique_keys(sample_df[unmatched_gt_before_col]) if unmatched_gt_before_col in sample_df.columns else []
    unmatched_pred_before_keys = _collect_unique_keys(sample_df[unmatched_pred_before_col]) if unmatched_pred_before_col in sample_df.columns else []

    soft_mask = (sample_df["soft_match_count"] > 0) if "soft_match_count" in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
    repaired_mask = (sample_df["pred_was_repaired"] > 0) if "pred_was_repaired" in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
    unmatched_gt_before_mask = sample_df[unmatched_gt_before_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_gt_before_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
    unmatched_pred_before_mask = sample_df[unmatched_pred_before_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_pred_before_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
    unmatched_gt_after_mask = sample_df[unmatched_gt_after_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_gt_after_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
    unmatched_pred_after_mask = sample_df[unmatched_pred_after_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_pred_after_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)

    return {
        "dimension_accuracy": float(component_stats.get("dimension_token_accuracy", 0.0)),
        "json_validity": float(mvss_summary.get("JSON Validity", 0.0)),
        "strict_attribute_match": float(component_stats.get("strict_attribute_match_accuracy", 0.0)),
        "MVSS_accuracy": float(mvss_summary.get("MVSS_accuracy", 0.0)),
        "MVSS_partial": float(mvss_summary.get("MVSS_partial", 0.0)),
        "total_value_count": int(component_stats.get("total_value_count", 0)),
        "correct_value_count": int(component_stats.get("correct_value_count", 0)),
        "value_accuracy": float(component_stats.get("value_accuracy", 0.0)),
        "total_unit_count": int(component_stats.get("total_unit_count", 0)),
        "correct_unit_count": int(component_stats.get("correct_unit_count", 0)),
        "unit_accuracy": float(component_stats.get("unit_accuracy", 0.0)),
        "total_source_view_count": int(component_stats.get("total_source_view_count", 0)),
        "correct_source_view_count": int(component_stats.get("correct_source_view_count", 0)),
        "source_view_accuracy": float(component_stats.get("source_view_accuracy", 0.0)),
        "total_null_count": int(component_stats.get("total_null_count", 0)),
        "correct_null_count": int(component_stats.get("correct_null_count", 0)),
        "null_accuracy": float(component_stats.get("null_accuracy", 0.0)),
        "total_attribute_count": int(component_stats.get("total_attribute_count", 0)),
        "strict_attribute_match_accuracy": float(component_stats.get("strict_attribute_match_accuracy", 0.0)),
        "gt_key_count": int(sample_df["gt_key_count"].sum()) if "gt_key_count" in sample_df.columns else 0,
        "pred_key_count": int(sample_df["pred_key_count"].sum()) if "pred_key_count" in sample_df.columns else 0,
        "matched_key_count_before_soft": int(sample_df["matched_key_count_before_soft"].sum()) if "matched_key_count_before_soft" in sample_df.columns else 0,
        "matched_key_count_after_soft": int(sample_df["matched_key_count_after_soft"].sum()) if "matched_key_count_after_soft" in sample_df.columns else 0,
        "matched_key_count": int(sample_df["matched_key_count"].sum()) if "matched_key_count" in sample_df.columns else 0,
        "extra_pred_key_count": int(sample_df["extra_pred_key_count"].sum()) if "extra_pred_key_count" in sample_df.columns else 0,
        "unmatched_gt_keys_before_soft": unmatched_gt_before_keys,
        "unmatched_pred_keys_before_soft": unmatched_pred_before_keys,
        "unmatched_gt_keys": unmatched_gt_after_keys,
        "unmatched_pred_keys": unmatched_pred_after_keys,
        "repaired_pred_count": int(sample_df["pred_was_repaired"].sum()) if "pred_was_repaired" in sample_df.columns else 0,
        "soft_matched_key_count": int(sample_df["soft_match_count"].sum()) if "soft_match_count" in sample_df.columns else 0,
        "unmatched_gt_key_count_before_soft": _sum_list_lengths(unmatched_gt_before_col),
        "unmatched_gt_key_count_after_soft": _sum_list_lengths(unmatched_gt_after_col),
        "unmatched_pred_key_count_before_soft": _sum_list_lengths(unmatched_pred_before_col),
        "unmatched_pred_key_count_after_soft": _sum_list_lengths(unmatched_pred_after_col),
        "unmatched_gt_question_numbers_before_soft": _collect_question_numbers(unmatched_gt_before_mask),
        "unmatched_pred_question_numbers_before_soft": _collect_question_numbers(unmatched_pred_before_mask),
        "unmatched_gt_question_numbers": _collect_question_numbers(unmatched_gt_after_mask),
        "unmatched_pred_question_numbers": _collect_question_numbers(unmatched_pred_after_mask),
        "soft_matched_question_numbers": _collect_question_numbers(soft_mask),
        "repaired_pred_question_numbers": _collect_question_numbers(repaired_mask),
    }

if "SKIP_CLAUDE" in globals() and SKIP_CLAUDE:
    print(f"{MODEL_NAME}: missing input CSV. Save dictionary cell skipped.")
    CLAUDE_METRICS = None
    CLAUDE_RELATION_GROUP_METRICS = None
    CLAUDE_RELATION_GROUP_TABLES = None
else:
    CLAUDE_METRICS = {}
    CLAUDE_METRICS.update(_group_sections(df, "layer_id", "LayerID"))
    CLAUDE_METRICS.update(_group_sections(df, "relation_type", "RelationType"))
    CLAUDE_METRICS.update(_group_sections(df, "entity_scope", "EntityScope"))
    CLAUDE_METRICS["Overall"] = _compute_metrics(df)

    if "_relation_type_norm" in df.columns:
        rel_norm = df["_relation_type_norm"].fillna("").astype(str)
    elif "relation_type" in df.columns:
        rel_norm = df["relation_type"].fillna("").astype(str).str.strip().str.lower()
    else:
        rel_norm = pd.Series([""] * len(df), index=df.index, dtype="object")
    mv_mask = rel_norm == "multi_view_dimensions"

    mv_component_stats, mvss_summary_raw, mv_sample_df, mv_attr_df = evaluate_mvss_dataframe(df[mv_mask], COL_GT, COL_PRED)
    mvss_summary = _build_mvss_overall_metrics(mv_component_stats, mvss_summary_raw, mv_sample_df)

    CLAUDE_RELATION_GROUP_METRICS = {
        "multi_view_dimensions_component": mv_component_stats,
        "multi_view_dimensions_mvss": mvss_summary,
    }

    CLAUDE_RELATION_GROUP_TABLES = {
        "multi_view_component_table": pd.DataFrame([mv_component_stats]),
        "mvss_summary_table": pd.DataFrame([mvss_summary]),
        "mvss_sample_table": mv_sample_df,
        "mvss_attr_table": mv_attr_df,
    }

    if "MODEL_METRICS" not in globals() or not isinstance(MODEL_METRICS, dict):
        MODEL_METRICS = {}
    MODEL_METRICS[MODEL_NAME] = CLAUDE_METRICS

CLAUDE_RELATION_GROUP_METRICS
````

````python
import pandas as pd
import re

#INPUT = "output_with_answers_recovered.csv"
INPUT = "1_2_with_answers_gemini.csv"
#INPUT = "1_1_1_Eval_Claude_Opus_1.csv"

COL_TEMPLATE = "template_id"
COL_GT = "ground_truth_answer"
COL_PRED = "model_answer"

from pathlib import Path
SKIP_GEMINI = not Path(INPUT).exists()
if SKIP_GEMINI:
    print(f"Gemini 3 Pro Preview: missing {INPUT}. Cell skipped. Dictionary not saved.")
    df = pd.DataFrame(columns=[COL_TEMPLATE, COL_GT, COL_PRED])
else:
    df = pd.read_csv(INPUT)

def normalize_text(s: str) -> str:
    s = "" if pd.isna(s) else str(s)
    s = s.lower().strip()
    s = re.sub(r"[^\w\s]", " ", s, flags=re.UNICODE)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def normalize_label(s: str) -> str:
    t = normalize_text(s)
    if t in {"yes", "true"}:
        return "yes"
    if t in {"no", "false"}:
        return "no"
    return t

def contains_whole_word(text: str, word: str) -> bool:
    if not word:
        return False
    pattern = r"(?:^|\s)" + re.escape(word) + r"(?:$|\s)"
    return re.search(pattern, text) is not None

df["gt_norm"] = df[COL_GT].apply(normalize_label)
df["pred_norm_text"] = df[COL_PRED].apply(normalize_text)

def predicted_label_from_text(pred_text: str, gt_label: str) -> str:
    return gt_label if contains_whole_word(pred_text, gt_label) else ""

df["pred_label"] = [
    predicted_label_from_text(p, g)
    for p, g in zip(df["pred_norm_text"], df["gt_norm"])
]

# correctness
df["is_correct"] = df["pred_label"] == df["gt_norm"]

# ---- metrics per each template_id ----
def group_metrics(g: pd.DataFrame) -> pd.Series:
    n = len(g)
    acc = g["is_correct"].mean() if n else 0.0

    labels = sorted(set(g["gt_norm"].unique()) | set(g["pred_label"].unique()))
    labels = [lab for lab in labels if lab != ""]  # remove empty label

    # F1:
    # - If only yes/no in GT -> binary F1 (positive=yes)
    # - Otherwise, macro F1 by GT classes (ignoring empty class)
    gt_set = set(g["gt_norm"].unique())
    if gt_set.issubset({"yes", "no"}):
        y_true = g["gt_norm"].tolist()
        y_pred = [("yes" if x == "yes" else "no") if x in {"yes","no"} else "no" for x in g["pred_label"].tolist()]
        # F1 positive=yes
        tp = sum((t == "yes") and (p == "yes") for t, p in zip(y_true, y_pred))
        fp = sum((t != "yes") and (p == "yes") for t, p in zip(y_true, y_pred))
        fn = sum((t == "yes") and (p != "yes") for t, p in zip(y_true, y_pred))
        precision = tp / (tp + fp) if (tp + fp) else 0.0
        recall = tp / (tp + fn) if (tp + fn) else 0.0
        f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0.0
        f1_type = "binary(yes)"
    else:
        # macro F1 per GT classes
        y_true = g["gt_norm"].tolist()
        y_pred = g["pred_label"].tolist()

        f1s = []
        for lab in sorted(gt_set):
            tp = sum((t == lab) and (p == lab) for t, p in zip(y_true, y_pred))
            fp = sum((t != lab) and (p == lab) for t, p in zip(y_true, y_pred))
            fn = sum((t == lab) and (p != lab) for t, p in zip(y_true, y_pred))
            precision = tp / (tp + fp) if (tp + fp) else 0.0
            recall = tp / (tp + fn) if (tp + fn) else 0.0
            f1_lab = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0.0
            f1s.append(f1_lab)
        f1 = sum(f1s) / len(f1s) if f1s else 0.0
        f1_type = "macro(GT-classes)"

    return pd.Series({
        "n": n,
        "accuracy": acc,
        "f1": f1,
        "f1_type": f1_type,
    })

rows = []
if SKIP_GEMINI:
    report = pd.DataFrame(columns=["n", "accuracy", "f1", "f1_type", COL_TEMPLATE])
else:
    for template_id, g in df.groupby(COL_TEMPLATE, dropna=False):
        m = group_metrics(g)          # Series: n, accuracy, f1, f1_type
        m[COL_TEMPLATE] = template_id
        rows.append(m)
    report = pd.DataFrame(rows).sort_values(COL_TEMPLATE).reset_index(drop=True)
report
````

````python
import pandas as pd

def _ensure_columns(df_in: pd.DataFrame) -> pd.DataFrame:
    out = df_in.copy()
    for c in ["layer_id", "relation_type", "entity_scope", "template_id", COL_GT, COL_PRED]:
        if c not in out.columns:
            out[c] = ""
    return out

if "SKIP_GPT" in globals() and SKIP_GPT:
    print("GPT-5.2: missing input CSV. Save dictionary cell skipped.")
    GPT_METRICS = None
    GPT_RELATION_GROUP_METRICS = None
    GPT_RELATION_GROUP_TABLES = None
else:
    df = _ensure_columns(df)

    def _group_sections(df_in: pd.DataFrame, col: str, prefix: str) -> dict:
        if col not in df_in.columns:
            return {}
        vals = df_in[col].fillna("").astype(str)
        sections = {}
        for v in sorted([v for v in vals.unique() if v != ""]):
            sections[f"{prefix}={v}"] = _compute_metrics(df_in[vals == v])
        return sections

    if "relation_type" in df.columns:
        rel = df["relation_type"].fillna("").astype(str).str.strip()
        rel_lower = rel.str.lower()
        rel = rel.where(~rel_lower.isin(["connectivity", "not_direct_connectivity"]), "connectivity")
        df["relation_type"] = rel

    if "_build_mvss_overall_metrics" not in globals():
        def _collect_unique_keys(series: pd.Series) -> list:
            out = []
            seen = set()
            for x in series.tolist():
                if isinstance(x, (list, tuple, set)):
                    vals = x
                elif pd.isna(x):
                    vals = []
                else:
                    vals = [x]
                for v in vals:
                    s = str(v).strip()
                    if s and s not in seen:
                        seen.add(s)
                        out.append(s)
            return out

        def _build_mvss_overall_metrics(component_stats: dict, mvss_summary: dict, sample_df: pd.DataFrame) -> dict:
            sample_df = sample_df if isinstance(sample_df, pd.DataFrame) else pd.DataFrame()

            def _as_list(v):
                if isinstance(v, list):
                    return v
                if isinstance(v, tuple):
                    return list(v)
                if isinstance(v, set):
                    return sorted(v)
                return []

            def _sum_list_lengths(col: str) -> int:
                if col not in sample_df.columns:
                    return 0
                return int(sample_df[col].apply(lambda x: len(_as_list(x))).sum())

            def _collect_question_numbers(mask: pd.Series) -> list:
                if "question_number" not in sample_df.columns:
                    return []
                out = []
                seen = set()
                for _, r in sample_df.loc[mask].iterrows():
                    q = r.get("question_number")
                    if pd.isna(q):
                        continue
                    if isinstance(q, float) and q.is_integer():
                        q = int(q)
                    qs = str(q).strip()
                    if qs and qs not in seen:
                        seen.add(qs)
                        out.append(qs)
                return out

            unmatched_gt_before_col = "unmatched_gt_keys_before_soft"
            unmatched_pred_before_col = "unmatched_pred_keys_before_soft"
            unmatched_gt_after_col = "unmatched_gt_keys_after_soft" if "unmatched_gt_keys_after_soft" in sample_df.columns else "unmatched_gt_keys"
            unmatched_pred_after_col = "unmatched_pred_keys_after_soft" if "unmatched_pred_keys_after_soft" in sample_df.columns else "unmatched_pred_keys"

            unmatched_gt_after_keys = _collect_unique_keys(sample_df[unmatched_gt_after_col]) if unmatched_gt_after_col in sample_df.columns else []
            unmatched_pred_after_keys = _collect_unique_keys(sample_df[unmatched_pred_after_col]) if unmatched_pred_after_col in sample_df.columns else []

            unmatched_gt_before_keys = _collect_unique_keys(sample_df[unmatched_gt_before_col]) if unmatched_gt_before_col in sample_df.columns else []
            unmatched_pred_before_keys = _collect_unique_keys(sample_df[unmatched_pred_before_col]) if unmatched_pred_before_col in sample_df.columns else []

            soft_mask = (sample_df["soft_match_count"] > 0) if "soft_match_count" in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
            repaired_mask = (sample_df["pred_was_repaired"] > 0) if "pred_was_repaired" in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
            unmatched_gt_before_mask = sample_df[unmatched_gt_before_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_gt_before_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
            unmatched_pred_before_mask = sample_df[unmatched_pred_before_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_pred_before_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
            unmatched_gt_after_mask = sample_df[unmatched_gt_after_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_gt_after_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)
            unmatched_pred_after_mask = sample_df[unmatched_pred_after_col].apply(lambda x: len(_as_list(x)) > 0) if unmatched_pred_after_col in sample_df.columns else pd.Series([False] * len(sample_df), index=sample_df.index)

            return {
                "dimension_accuracy": float(component_stats.get("dimension_token_accuracy", 0.0)),
                "json_validity": float(mvss_summary.get("JSON Validity", 0.0)),
                "strict_attribute_match": float(component_stats.get("strict_attribute_match_accuracy", 0.0)),
                "MVSS_accuracy": float(mvss_summary.get("MVSS_accuracy", 0.0)),
                "MVSS_partial": float(mvss_summary.get("MVSS_partial", 0.0)),
                "total_value_count": int(component_stats.get("total_value_count", 0)),
                "correct_value_count": int(component_stats.get("correct_value_count", 0)),
                "value_accuracy": float(component_stats.get("value_accuracy", 0.0)),
                "total_unit_count": int(component_stats.get("total_unit_count", 0)),
                "correct_unit_count": int(component_stats.get("correct_unit_count", 0)),
                "unit_accuracy": float(component_stats.get("unit_accuracy", 0.0)),
                "total_source_view_count": int(component_stats.get("total_source_view_count", 0)),
                "correct_source_view_count": int(component_stats.get("correct_source_view_count", 0)),
                "source_view_accuracy": float(component_stats.get("source_view_accuracy", 0.0)),
                "total_null_count": int(component_stats.get("total_null_count", 0)),
                "correct_null_count": int(component_stats.get("correct_null_count", 0)),
                "null_accuracy": float(component_stats.get("null_accuracy", 0.0)),
                "total_attribute_count": int(component_stats.get("total_attribute_count", 0)),
                "strict_attribute_match_accuracy": float(component_stats.get("strict_attribute_match_accuracy", 0.0)),
                "gt_key_count": int(sample_df["gt_key_count"].sum()) if "gt_key_count" in sample_df.columns else 0,
                "pred_key_count": int(sample_df["pred_key_count"].sum()) if "pred_key_count" in sample_df.columns else 0,
                "matched_key_count_before_soft": int(sample_df["matched_key_count_before_soft"].sum()) if "matched_key_count_before_soft" in sample_df.columns else 0,
                "matched_key_count_after_soft": int(sample_df["matched_key_count_after_soft"].sum()) if "matched_key_count_after_soft" in sample_df.columns else 0,
                "matched_key_count": int(sample_df["matched_key_count"].sum()) if "matched_key_count" in sample_df.columns else 0,
                "extra_pred_key_count": int(sample_df["extra_pred_key_count"].sum()) if "extra_pred_key_count" in sample_df.columns else 0,
                "unmatched_gt_keys_before_soft": unmatched_gt_before_keys,
                "unmatched_pred_keys_before_soft": unmatched_pred_before_keys,
                "unmatched_gt_keys": unmatched_gt_after_keys,
                "unmatched_pred_keys": unmatched_pred_after_keys,
                "repaired_pred_count": int(sample_df["pred_was_repaired"].sum()) if "pred_was_repaired" in sample_df.columns else 0,
                "soft_matched_key_count": int(sample_df["soft_match_count"].sum()) if "soft_match_count" in sample_df.columns else 0,
                "unmatched_gt_key_count_before_soft": _sum_list_lengths(unmatched_gt_before_col),
                "unmatched_gt_key_count_after_soft": _sum_list_lengths(unmatched_gt_after_col),
                "unmatched_pred_key_count_before_soft": _sum_list_lengths(unmatched_pred_before_col),
                "unmatched_pred_key_count_after_soft": _sum_list_lengths(unmatched_pred_after_col),
                "unmatched_gt_question_numbers_before_soft": _collect_question_numbers(unmatched_gt_before_mask),
                "unmatched_pred_question_numbers_before_soft": _collect_question_numbers(unmatched_pred_before_mask),
                "unmatched_gt_question_numbers": _collect_question_numbers(unmatched_gt_after_mask),
                "unmatched_pred_question_numbers": _collect_question_numbers(unmatched_pred_after_mask),
                "soft_matched_question_numbers": _collect_question_numbers(soft_mask),
                "repaired_pred_question_numbers": _collect_question_numbers(repaired_mask),
            }

    GPT_METRICS = {}
    GPT_METRICS.update(_group_sections(df, "layer_id", "LayerID"))
    GPT_METRICS.update(_group_sections(df, "relation_type", "RelationType"))
    GPT_METRICS.update(_group_sections(df, "entity_scope", "EntityScope"))
    GPT_METRICS["Overall"] = _compute_metrics(df)

    if "_relation_type_norm" in df.columns:
        rel_norm = df["_relation_type_norm"].fillna("").astype(str)
    elif "relation_type" in df.columns:
        rel_norm = df["relation_type"].fillna("").astype(str).str.strip().str.lower()
    else:
        rel_norm = pd.Series([""] * len(df), index=df.index, dtype="object")
    mv_mask = rel_norm == "multi_view_dimensions"

    mv_component_stats, mvss_summary_raw, mv_sample_df, mv_attr_df = evaluate_mvss_dataframe(df[mv_mask], COL_GT, COL_PRED)
    mvss_summary = _build_mvss_overall_metrics(mv_component_stats, mvss_summary_raw, mv_sample_df)

    GPT_RELATION_GROUP_METRICS = {
        "multi_view_dimensions_component": mv_component_stats,
        "multi_view_dimensions_mvss": mvss_summary,
    }
    GPT_RELATION_GROUP_TABLES = {
        "multi_view_component_table": pd.DataFrame([mv_component_stats]),
        "mvss_summary_table": pd.DataFrame([mvss_summary]),
        "mvss_sample_table": mv_sample_df,
        "mvss_attr_table": mv_attr_df,
    }

    if "MODEL_METRICS" not in globals() or not isinstance(MODEL_METRICS, dict):
        MODEL_METRICS = {}
    MODEL_METRICS["GPT-5.2"] = GPT_METRICS

GPT_RELATION_GROUP_METRICS
````

# 5. Final Summary Table

Combines model dictionaries and exports the final MVSS summary.
Main output:
- `1_2_mvss_summary.csv`


````python
import json
import pandas as pd

OUTPUT_SUMMARY_CSV = "1_2_mvss_summary.csv"

MODEL_COLUMNS = [
    "Claude Opus 4.6",
    "Gemini 3 Pro Preview",
    "GPT-5.2",
    "GPT-5.4",
]

MODEL_VAR_CANDIDATES = {
    "Claude Opus 4.6": ["CLAUDE_RELATION_GROUP_METRICS"],
    "Gemini 3 Pro Preview": ["GEMINI_RELATION_GROUP_METRICS"],
    "GPT-5.2": ["GPT_RELATION_GROUP_METRICS"],
    "GPT-5.4": ["GPT54_RELATION_GROUP_METRICS", "GPT_5_4_RELATION_GROUP_METRICS"],
}

METRIC_ORDER = [
    "dimension_accuracy",
    "json_validity",
    "strict_attribute_match",
    "MVSS_accuracy",
    "MVSS_partial",
    "repaired_pred_count",
    "soft_matched_key_count",
    "unmatched_gt_key_count_before_soft",
    "unmatched_gt_key_count_after_soft",
    "unmatched_pred_key_count_before_soft",
    "unmatched_pred_key_count_after_soft",
    "total_value_count",
    "correct_value_count",
    "value_accuracy",
    "total_unit_count",
    "correct_unit_count",
    "unit_accuracy",
    "total_source_view_count",
    "correct_source_view_count",
    "source_view_accuracy",
    "total_null_count",
    "correct_null_count",
    "null_accuracy",
    "total_attribute_count",
    "strict_attribute_match_accuracy",
    "gt_key_count",
    "pred_key_count",
    "matched_key_count_before_soft",
    "matched_key_count_after_soft",
    "matched_key_count",
    "extra_pred_key_count",
    "unmatched_gt_keys_before_soft",
    "unmatched_pred_keys_before_soft",
    "unmatched_gt_keys",
    "unmatched_pred_keys",
    "soft_matched_question_numbers",
    "repaired_pred_question_numbers",
    "unmatched_gt_question_numbers_before_soft",
    "unmatched_pred_question_numbers_before_soft",
    "unmatched_gt_question_numbers",
    "unmatched_pred_question_numbers",
]

def _get_model_relation_group_metrics(model_name: str):
    for var_name in MODEL_VAR_CANDIDATES.get(model_name, []):
        if var_name in globals() and isinstance(globals()[var_name], dict):
            return globals()[var_name]
    return None

def _fmt_value(v):
    if isinstance(v, (list, tuple, dict)):
        return json.dumps(v, ensure_ascii=False)
    return v

rows = []
for metric in METRIC_ORDER:
    row = {"Metric": metric}
    for model_name in MODEL_COLUMNS:
        rel_metrics = _get_model_relation_group_metrics(model_name)
        mvss_block = rel_metrics.get("multi_view_dimensions_mvss") if isinstance(rel_metrics, dict) else None
        value = mvss_block.get(metric, "") if isinstance(mvss_block, dict) else ""
        row[model_name] = _fmt_value(value)
    rows.append(row)

summary_df = pd.DataFrame(rows, columns=["Metric"] + MODEL_COLUMNS)
summary_df.to_csv(OUTPUT_SUMMARY_CSV, index=False, encoding="utf-8-sig")
print("Saved:", OUTPUT_SUMMARY_CSV)
summary_df
````

### Generate Excel

Reads `1_2_mvss_summary.csv` and creates a formatted Excel report:
- `1_2_mvss_summary.xlsx`


````python
import json
import pandas as pd
from pathlib import Path

csv_path = Path("1_2_mvss_summary.csv")
xlsx_path = csv_path.with_suffix(".xlsx")

summary_df = pd.read_csv(csv_path)

MODEL_EXPORTS = {
    "Claude Opus 4.6": {
        "metrics_vars": ["CLAUDE_RELATION_GROUP_METRICS"],
        "tables_vars": ["CLAUDE_RELATION_GROUP_TABLES"],
        "slug": "claude_opus_4_6",
    },
    "Gemini 3 Pro Preview": {
        "metrics_vars": ["GEMINI_RELATION_GROUP_METRICS"],
        "tables_vars": ["GEMINI_RELATION_GROUP_TABLES"],
        "slug": "gemini_3_pro",
    },
    "GPT-5.2": {
        "metrics_vars": ["GPT_RELATION_GROUP_METRICS"],
        "tables_vars": ["GPT_RELATION_GROUP_TABLES"],
        "slug": "gpt_5_2",
    },
    "GPT-5.4": {
        "metrics_vars": ["GPT54_RELATION_GROUP_METRICS", "GPT_5_4_RELATION_GROUP_METRICS"],
        "tables_vars": ["GPT54_RELATION_GROUP_TABLES", "GPT_5_4_RELATION_GROUP_TABLES"],
        "slug": "gpt_5_4",
    },
}

def _get_first_dict(var_names):
    for name in var_names:
        if name in globals() and isinstance(globals()[name], dict):
            return globals()[name]
    return None

def _fmt_value(v):
    if isinstance(v, (list, tuple, dict)):
        return json.dumps(v, ensure_ascii=False)
    return v

with pd.ExcelWriter(xlsx_path, engine="openpyxl") as writer:
    summary_df.to_excel(writer, sheet_name="summary", index=False)

    list_rows = []
    sanity_rows = []

    for model_name, cfg in MODEL_EXPORTS.items():
        rel_metrics = _get_first_dict(cfg["metrics_vars"])
        rel_tables = _get_first_dict(cfg["tables_vars"])
        if not isinstance(rel_metrics, dict):
            continue

        mvss = rel_metrics.get("multi_view_dimensions_mvss", {})
        if isinstance(mvss, dict):
            sanity_rows.append({
                "model": model_name,
                "JSON Validity": mvss.get("json_validity", ""),
                "Strict Attribute Match": mvss.get("strict_attribute_match", ""),
                "MVSS_accuracy": mvss.get("MVSS_accuracy", ""),
                "MVSS_partial": mvss.get("MVSS_partial", ""),
                "repaired_pred_count": mvss.get("repaired_pred_count", ""),
                "soft_matched_key_count": mvss.get("soft_matched_key_count", ""),
            })
            list_rows.append({
                "model": model_name,
                "soft_matched_question_numbers": _fmt_value(mvss.get("soft_matched_question_numbers", [])),
                "repaired_pred_question_numbers": _fmt_value(mvss.get("repaired_pred_question_numbers", [])),
                "unmatched_gt_question_numbers_before_soft": _fmt_value(mvss.get("unmatched_gt_question_numbers_before_soft", [])),
                "unmatched_gt_question_numbers": _fmt_value(mvss.get("unmatched_gt_question_numbers", [])),
                "unmatched_pred_question_numbers_before_soft": _fmt_value(mvss.get("unmatched_pred_question_numbers_before_soft", [])),
                "unmatched_pred_question_numbers": _fmt_value(mvss.get("unmatched_pred_question_numbers", [])),
            })

        if not isinstance(rel_tables, dict):
            continue

        sample_df = rel_tables.get("mvss_sample_table")
        attr_df = rel_tables.get("mvss_attr_table")

        if isinstance(sample_df, pd.DataFrame) and len(sample_df):
            sample_out = sample_df.copy()
            for c in sample_out.columns:
                sample_out[c] = sample_out[c].apply(_fmt_value)
            sample_out.to_excel(writer, sheet_name=f"{cfg['slug']}_sample"[:31], index=False)

        if isinstance(attr_df, pd.DataFrame) and len(attr_df):
            attr_out = attr_df.copy()
            if isinstance(sample_df, pd.DataFrame) and len(sample_df):
                if "pred_was_repaired" not in attr_out.columns and "pred_was_repaired" in sample_df.columns:
                    attr_out = attr_out.merge(
                        sample_df[["row_index", "pred_was_repaired"]].drop_duplicates(),
                        on="row_index",
                        how="left",
                    )
            keep_cols = [
                "row_index",
                "question_number",
                "gt_key",
                "pred_key",
                "match_mode",
                "was_soft_matched",
                "partial_credit",
                "strict_match",
                "strict_match_base",
                "pred_was_repaired",
                "dimension_match",
                "value_match",
                "unit_match",
                "source_match",
                "null_match",
            ]
            keep_cols = [c for c in keep_cols if c in attr_out.columns]
            attr_out = attr_out[keep_cols]
            for c in attr_out.columns:
                attr_out[c] = attr_out[c].apply(_fmt_value)
            attr_out.to_excel(writer, sheet_name=f"{cfg['slug']}_details"[:31], index=False)

    if list_rows:
        pd.DataFrame(list_rows).to_excel(writer, sheet_name="question_lists", index=False)

    if sanity_rows:
        sanity_df = pd.DataFrame(sanity_rows)
        sanity_df.to_excel(writer, sheet_name="sanity_check", index=False)
        print("Sanity-check by model:")
        display(sanity_df)

print(f"Excel file created: {xlsx_path}")
````
