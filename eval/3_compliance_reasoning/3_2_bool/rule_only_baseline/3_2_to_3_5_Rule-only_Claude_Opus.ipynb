## 3. OpenRouter call utilities

```python
import base64
import json
import mimetypes
import time
from typing import Any, Dict, Optional

import requests


def build_headers(api_key: str) -> Dict[str, str]:
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json",
    }
    return headers


def check_key_info(api_key: str) -> Dict[str, Any]:
    """GET /key ? check usage/credits for the key."""
    r = requests.get(f"{OPENROUTER_BASE_URL}/key", headers=build_headers(api_key), timeout=30)
    r.raise_for_status()
    return r.json()

def build_prompt(question_text: str) -> str:
    q = "" if question_text is None else str(question_text).strip()
    return q

def normalize_image_value(v: Any) -> Optional[str]:
    if v is None:
        return None
    try:
        import pandas as _pd
        if _pd.isna(v):
            return None
    except Exception:
        pass
    s = str(v).strip().strip('"').strip("'")
    return s if s else None

def collect_image_values(*values: Any) -> list[str]:
    out: list[str] = []
    seen = set()
    for v in values:
        s = normalize_image_value(v)
        if not s:
            continue
        if s in seen:
            continue
        out.append(s)
        seen.add(s)
    return out

def is_image_ref(s: str) -> bool:
    if not s:
        return False
    return Path(s).exists()

def to_image_url(image_value: str) -> str:
    s = "" if image_value is None else str(image_value).strip().strip('"').strip("'")
    if not s:
        raise ValueError("Empty image value")
    p = Path(s)
    if not p.exists():
        raise FileNotFoundError(s)
    mime, _ = mimetypes.guess_type(p.name)
    if not mime:
        mime = "image/jpeg"
    data = p.read_bytes()
    b64 = base64.b64encode(data).decode("ascii")
    return f"data:{mime};base64,{b64}"

def call_openrouter_chat(api_key: str, prompt: str, image_values: Optional[list[str]], request_id: str | None = None) -> dict:
    url = f"{OPENROUTER_BASE_URL}/chat/completions"

    prompt_final = build_prompt(prompt)
    content = [{"type": "text", "text": prompt_final}]
    for image_value in (image_values or []):
        if not image_value:
            continue
        if not is_image_ref(image_value):
            continue
        image_url = to_image_url(image_value)
        content.append({"type": "image_url", "image_url": {"url": image_url}})

    if INSTRUCTION_SUFFIX:
        content.append({"type": "text", "text": INSTRUCTION_SUFFIX})

    payload = {
        "model": MODEL,
        "messages": [
            {
                "role": "user",
                "content": content,
            }
        ],
        "max_tokens": MAX_TOKENS,
        "temperature": TEMPERATURE,
    }

    if request_id is not None:
        payload["metadata"] = {"request_id": request_id}

    r = requests.post(url, headers=build_headers(api_key), data=json.dumps(payload), timeout=120)
    if r.status_code >= 400:
        try:
            err = r.json()
        except Exception:
            err = {"error": {"message": r.text}}
        raise RuntimeError(f"OpenRouter error {r.status_code}: {err}")

    return r.json()



def call_with_retries(api_key: str, prompt: str, image_values: Optional[list[str]], request_id: str, max_retries: int = 6) -> Dict[str, Any]:
    """Retries for transient errors (429/502/503/408)."""
    backoff = 2.0
    for attempt in range(max_retries + 1):
        try:
            return call_openrouter_chat(api_key, prompt, image_values, request_id=request_id)
        except RuntimeError as e:
            msg = str(e)
            retriable = any(code in msg for code in [" 429:", " 502:", " 503:", " 408:"])
            if (attempt == max_retries) or (not retriable):
                raise
            time.sleep(backoff)
            backoff *= 1.7
    raise RuntimeError("Unreachable")


def extract_text(resp: Dict[str, Any]) -> str:
    """choices[0].message.content"""
    try:
        return resp["choices"][0]["message"]["content"]
    except Exception:
        return ""
```

### Run the dataset on the model:

```python
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import json

INPUT_CSV = "C:\Users\ritaMZ\PycharmProjects\bravo_benchmark\eval\3_compliance_reasoning\3_2_bool\rule_only_baseline\3_2_to_3_5_(2)_rule_onl_subset.csv"
CLEAN_CSV = "3_2_rule_only_with_answers_claude_opus_4_6_merged80.csv"
PREVIEW_CSV = "3_2_rule_only_preview_claude_opus_4_6.csv"

df_input = pd.read_csv(INPUT_CSV)

assert QUESTION_COL in df_input.columns, f"CSV must contain column '{QUESTION_COL}'. Found: {list(df_input.columns)}"
assert IMAGE_COL in df_input.columns, f"CSV must contain column '{IMAGE_COL}'. Found: {list(df_input.columns)}"
assert "generated_question_id" in df_input.columns, "CSV must contain generated_question_id."

def answered_mask(series: pd.Series) -> pd.Series:
    s = series.fillna("").astype(str).str.strip().str.lower()
    return ~s.isin(["", "0", "nan"])

answer_cols = ["generated_question_id", "model", "openrouter_request_id", "model_answer", "error_reason"]

existing_parts = []

if Path(CLEAN_CSV).exists():
    clean_df = pd.read_csv(CLEAN_CSV)
    if "generated_question_id" in clean_df.columns:
        for col in answer_cols[1:]:
            if col not in clean_df.columns:
                clean_df[col] = ""
        clean_df["_source_priority"] = 2
        existing_parts.append(clean_df[answer_cols + ["_source_priority"]])
        print(f"Loaded {len(clean_df)} rows from clean CSV.")
    else:
        print("Clean CSV found but missing generated_question_id; skipping.")

if Path(PREVIEW_CSV).exists():
    preview_df = pd.read_csv(PREVIEW_CSV)
    if "generated_question_id" in preview_df.columns:
        for col in answer_cols[1:]:
            if col not in preview_df.columns:
                preview_df[col] = ""
        preview_df["model"] = preview_df["model"].replace("", MODEL).fillna(MODEL)
        preview_df["_source_priority"] = 1
        existing_parts.append(preview_df[answer_cols + ["_source_priority"]])
        print(f"Loaded {len(preview_df)} rows from preview CSV.")
    else:
        print("Preview CSV found but missing generated_question_id; skipping.")

base_df = df_input.drop(columns=[c for c in answer_cols[1:] if c in df_input.columns], errors="ignore")

if existing_parts:
    existing_df = pd.concat(existing_parts, ignore_index=True)
    existing_df["model_answer"] = existing_df["model_answer"].fillna("")
    existing_df["_answered"] = answered_mask(existing_df["model_answer"])
    existing_best = (
        existing_df.sort_values(
            ["generated_question_id", "_answered", "_source_priority"],
            ascending=[True, False, False],
        )
        .drop_duplicates(subset=["generated_question_id"], keep="first")
    )
    df = base_df.merge(existing_best[answer_cols], on="generated_question_id", how="left")
else:
    df = base_df.copy()
    for col in answer_cols[1:]:
        df[col] = ""

df["model_answer"] = df["model_answer"].fillna("")
df["error_reason"] = df["error_reason"].fillna("")
answered_before = answered_mask(df["model_answer"])
df_pending = df[~answered_before].copy()

print(
    f"Total rows={len(df)} | already answered={int(answered_before.sum())} | pending={len(df_pending)}"
)

answers = []
req_ids = []
errors = []

with open(OUTPUT_JSONL, "w", encoding="utf-8") as fjsonl:
    for i, row in tqdm(df_pending.iterrows(), total=len(df_pending)):
        prompt = str(row[QUESTION_COL])
        image_values = collect_image_values(row.get(IMAGE_COL))
        request_id = f"row_{i}"
        generated_question_id = row.get("generated_question_id")
        error_reason = ""

        try:
            resp = call_with_retries(OPENROUTER_API_KEY, prompt, image_values, request_id)
            text = extract_text(resp)
            fjsonl.write(json.dumps(
                {"row_index": i, "generated_question_id": generated_question_id, "request_id": request_id, "response": resp},
                ensure_ascii=False
            ) + "\\n")
        except RuntimeError as e:
            text = ""
            error_reason = str(e)
            fjsonl.write(json.dumps(
                {"row_index": i, "generated_question_id": generated_question_id, "request_id": request_id, "error": error_reason},
                ensure_ascii=False
            ) + "\\n")

        answers.append(text)
        req_ids.append(request_id)
        errors.append(error_reason)

run_df = df_pending.copy()
run_df["model"] = MODEL
run_df["openrouter_request_id"] = req_ids
run_df["model_answer"] = answers
run_df["error_reason"] = errors

if not run_df.empty:
    run_update = (
        run_df[["generated_question_id", "model", "openrouter_request_id", "model_answer", "error_reason"]]
        .drop_duplicates(subset=["generated_question_id"], keep="last")
        .set_index("generated_question_id")
    )
    df = df.set_index("generated_question_id")
    for col in ["model", "openrouter_request_id", "model_answer", "error_reason"]:
        df.loc[run_update.index, col] = run_update[col]
    df = df.reset_index()

df["model"] = df["model"].fillna("").replace("", MODEL)
df["openrouter_request_id"] = df["openrouter_request_id"].fillna("")
df["model_answer"] = df["model_answer"].fillna("")
df["error_reason"] = df["error_reason"].fillna("")

run_df.to_csv(OUTPUT_CSV, index=False)

append_cols = ["generated_question_id", "model", "openrouter_request_id", "model_answer", "error_reason"]
append_df = run_df[append_cols].copy()

if Path(CLEAN_CSV).exists():
    append_df.to_csv(CLEAN_CSV, mode="a", header=False, index=False)
    print(f"Appended {len(append_df)} rows to existing clean CSV: {CLEAN_CSV}")
else:
    append_df.to_csv(CLEAN_CSV, index=False)
    print(f"Created clean CSV with {len(append_df)} rows: {CLEAN_CSV}")

print("Saved run batch:", OUTPUT_CSV)
print("Saved raw log:", OUTPUT_JSONL)
print("Remaining empty answers in current merged view:", int((~answered_mask(df["model_answer"])).sum()))

df.head(3)
```

# 5. Metrics

```python
import json
from pathlib import Path
import pandas as pd

JSONL_PATH = "../results/3_2_rule_only_with_answers_claude_opus_4_6_raw(1).jsonl"
QUESTIONS_CSV = "C:\Users\ritaMZ\PycharmProjects\bravo_benchmark\eval\3_compliance_reasoning\3_2_bool\rule_only_baseline\3_2_to_3_5_(2)_rule_onl_subset.csv"
OUT_CSV = "3_2_rule_only_clean_claude_opus_4_6.csv"
MODEL_CSV = "3_2_rule_only_with_answers_claude_opus_4_6.csv"
PREVIEW_CSV = "3_2_rule_only_preview_claude_opus_4_6.csv"

KEEP_COLS = ["model_answer", "openrouter_request_id", "error_reason"]

def is_empty(series: pd.Series) -> pd.Series:
    s = series.astype(str).str.strip()
    return series.isna() | (s == "") | (s.str.lower() == "nan")

def load_jsonl(path: str) -> pd.DataFrame:
    rows = []
    if Path(path).exists():
        with open(path, "r", encoding="utf-8") as f:
            for line in f:
                rec = json.loads(line)
                row_index = rec.get("row_index")
                generated_question_id = rec.get("generated_question_id")
                request_id = rec.get("request_id")
                error_reason = rec.get("error")
                resp = rec.get("response", {}) or {}

                text = ""
                if not error_reason:
                    try:
                        text = resp["choices"][0]["message"]["content"]
                    except Exception:
                        text = ""

                rows.append({
                    "row_index": row_index,
                    "generated_question_id": generated_question_id,
                    "openrouter_request_id": request_id,
                    "model_answer": text,
                    "error_reason": error_reason,
                })
    return pd.DataFrame(rows)

def merge_src(df, src, name):
    if src is None or src.empty:
        return df
    if "generated_question_id" not in src.columns:
        print(f"WARNING: {name} missing generated_question_id; skipping")
        return df
    src = src.copy()
    for c in KEEP_COLS:
        if c not in src.columns:
            src[c] = ""
    src = src[["generated_question_id"] + KEEP_COLS]
    out = df.merge(src, on="generated_question_id", how="left", suffixes=("", f"_{name}"))
    for c in KEEP_COLS:
        out[c] = out[c].where(~is_empty(out[c]), out[f"{c}_{name}"])
        out.drop(columns=[f"{c}_{name}"], inplace=True)
    return out

# 1) dataset is the source of truth for metadata
df_base = pd.read_csv(QUESTIONS_CSV)
if "generated_question_id" not in df_base.columns:
    raise ValueError("Dataset must contain generated_question_id")
for c in KEEP_COLS:
    if c not in df_base.columns:
        df_base[c] = ""

# 2) load sources
df_existing = pd.read_csv(OUT_CSV) if Path(OUT_CSV).exists() else None
df_model = pd.read_csv(MODEL_CSV) if Path(MODEL_CSV).exists() else None
df_json = load_jsonl(JSONL_PATH) if Path(JSONL_PATH).exists() else None
df_preview = pd.read_csv(PREVIEW_CSV) if Path(PREVIEW_CSV).exists() else None

# 3) merge model outputs by generated_question_id only
df_out = df_base.copy()
df_out = merge_src(df_out, df_existing, "existing")
df_out = merge_src(df_out, df_model, "model")
df_out = merge_src(df_out, df_json, "json")
df_out = merge_src(df_out, df_preview, "preview")

df_out = df_out.sort_values("generated_question_id").reset_index(drop=True)
df_out["row_index"] = df_out.index

df_out.to_csv(OUT_CSV, index=False)
print("Saved:", OUT_CSV)
df_out.head()
```

### Extract Yes/No answers:

```python
import pandas as pd
import re

INPUT = "3_2_rule_only_clean_claude_opus_4_6.csv"

df = pd.read_csv(INPUT)

def extract_yes_no(text: str) -> str:
    s = "" if pd.isna(text) else str(text)
    # Find explicit @YES/@NO marker, prefer the last one if multiple
    matches = re.findall(r"@\s*(YES|NO)\b", s, flags=re.IGNORECASE)
    if not matches:
        return ""
    return matches[-1].lower()

df["bool_model_answer"] = df["model_answer"].apply(extract_yes_no)

df.to_csv(INPUT, index=False)
print("Updated:", INPUT)
df[["model_answer", "bool_model_answer"]].head()
```

### F1&Accuracy

```python
import pandas as pd
import re

#INPUT = "output_with_answers_recovered.csv"
INPUT = "3_2_rule_only_clean_claude_opus_4_6.csv"
#INPUT = "1_1_1_Eval_Claude_Opus_1.csv"

COL_TEMPLATE = "template_id"
COL_GT = "correct_answer"
COL_PRED = "model_answer"

df = pd.read_csv(INPUT)

def normalize_text(s: str) -> str:
    s = "" if pd.isna(s) else str(s)
    s = s.lower().strip()
    s = re.sub(r"[^\w\s]", " ", s, flags=re.UNICODE)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def normalize_label(s: str) -> str:
    t = normalize_text(s)
    if t in {"yes", "true"}:
        return "yes"
    if t in {"no", "false"}:
        return "no"
    return t

def contains_whole_word(text: str, word: str) -> bool:
    if not word:
        return False
    pattern = r"(?:^|\s)" + re.escape(word) + r"(?:$|\s)"
    return re.search(pattern, text) is not None

df["gt_norm"] = df[COL_GT].apply(normalize_label)
df["pred_norm_text"] = df[COL_PRED].apply(normalize_text)

def predicted_label_from_text(pred_text: str, gt_label: str) -> str:
    return gt_label if contains_whole_word(pred_text, gt_label) else ""

df["pred_label"] = [
    predicted_label_from_text(p, g)
    for p, g in zip(df["pred_norm_text"], df["gt_norm"])
]

# correctness
df["is_correct"] = df["pred_label"] == df["gt_norm"]

# ---- metrics per each template_id ----
def group_metrics(g: pd.DataFrame) -> pd.Series:
    n = len(g)
    acc = g["is_correct"].mean() if n else 0.0

    labels = sorted(set(g["gt_norm"].unique()) | set(g["pred_label"].unique()))
    labels = [lab for lab in labels if lab != ""]  # СѓР±РµСЂС‘Рј РїСѓСЃС‚РѕР№

    # F1:
    # - If only yes/no in GT -> binary F1 (positive=yes)
    # - Otherwise, macro F1 by GT classes (ignoring empty class)
    gt_set = set(g["gt_norm"].unique())
    if gt_set.issubset({"yes", "no"}):
        y_true = g["gt_norm"].tolist()
        y_pred = [("yes" if x == "yes" else "no") if x in {"yes","no"} else "no" for x in g["pred_label"].tolist()]
        # F1 positive=yes
        tp = sum((t == "yes") and (p == "yes") for t, p in zip(y_true, y_pred))
        fp = sum((t != "yes") and (p == "yes") for t, p in zip(y_true, y_pred))
        fn = sum((t == "yes") and (p != "yes") for t, p in zip(y_true, y_pred))
        precision = tp / (tp + fp) if (tp + fp) else 0.0
        recall = tp / (tp + fn) if (tp + fn) else 0.0
        f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0.0
        f1_type = "binary(yes)"
    else:
        # macro F1 per GT classes
        y_true = g["gt_norm"].tolist()
        y_pred = g["pred_label"].tolist()

        f1s = []
        for lab in sorted(gt_set):
            tp = sum((t == lab) and (p == lab) for t, p in zip(y_true, y_pred))
            fp = sum((t != lab) and (p == lab) for t, p in zip(y_true, y_pred))
            fn = sum((t == lab) and (p != lab) for t, p in zip(y_true, y_pred))
            precision = tp / (tp + fp) if (tp + fp) else 0.0
            recall = tp / (tp + fn) if (tp + fn) else 0.0
            f1_lab = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0.0
            f1s.append(f1_lab)
        f1 = sum(f1s) / len(f1s) if f1s else 0.0
        f1_type = "macro(GT-classes)"

    return pd.Series({
        "n": n,
        "accuracy": acc,
        "f1": f1,
        "f1_type": f1_type,
    })

rows = []
for template_id, g in df.groupby(COL_TEMPLATE, dropna=False):
    m = group_metrics(g)          # Series: n, accuracy, f1, f1_type
    m[COL_TEMPLATE] = template_id
    rows.append(m)

report = pd.DataFrame(rows).sort_values(COL_TEMPLATE).reset_index(drop=True)
report
```

```python
# Summary metrics by Ambiguity + overall
import numpy as np
from sklearn.metrics import f1_score

def summarize_subset(df_in: pd.DataFrame, label: str):
    sub = df_in.copy()
    n = len(sub)
    acc = sub["is_correct"].mean() if n else 0.0
    f1 = group_metrics(sub)["f1"] if n else 0.0

    # Label distribution (by ground-truth)
    dist = sub["gt_norm"].value_counts(dropna=False)
    dist_str = ", ".join([f"{k}:{v}" for k, v in dist.items()]) if n else ""

    # Majority baseline (by ground-truth)
    majority_count = int(dist.max()) if n else 0
    majority_acc = (majority_count / n) if n else 0.0

    # Validity: prediction is in the answer scope for the question type
    if n:
        pred_valid = sub["bool_model_answer"].fillna("").astype(str).str.strip().str.lower()
        if sub["gt_norm"].isin(["yes", "no"]).all():
            # binary scope
            valid = pred_valid.isin(["yes", "no"])
        else:
            # fallback scope: use observed GT labels in this subset
            scope = set(sub["gt_norm"].dropna().unique())
            valid = pred_valid.isin(scope) if scope else (pred_valid != "")
    else:
        valid = pd.Series([], dtype=bool)
    validity = valid.mean() if n else 0.0

    # F1 breakdown for binary yes/no (treat non-yes as no)
    if n:
        y_true = (sub["gt_norm"] == "yes").astype(int).to_numpy()
        y_pred = (sub["pred_label"] == "yes").astype(int).to_numpy()
        f1_macro = f1_score(y_true, y_pred, average="macro")
        f1_yes = f1_score(y_true, y_pred, pos_label=1)
        f1_no = f1_score(y_true, y_pred, pos_label=0)

        maj_label = sub["gt_norm"].value_counts().idxmax()
        y_base = np.ones_like(y_true) if maj_label == "yes" else np.zeros_like(y_true)
        f1_macro_base = f1_score(y_true, y_base, average="macro")
        f1_yes_base = f1_score(y_true, y_base, pos_label=1)
        f1_no_base = f1_score(y_true, y_base, pos_label=0)
    else:
        f1_macro = f1_yes = f1_no = 0.0
        f1_macro_base = f1_yes_base = f1_no_base = 0.0

    print(
        f"{label}: n={n}, accuracy={acc:.4f}, f1={f1:.4f}, "
        f"f1_macro={f1_macro:.4f}, f1_yes={f1_yes:.4f}, f1_no={f1_no:.4f}, "
        f"majority_baseline={majority_acc:.4f}, validity={validity:.4f}"
    )
    if n:
        print(f"  label_distribution: {dist_str}")
        print(
            f"  baseline_f1: macro={f1_macro_base:.4f}, yes={f1_yes_base:.4f}, no={f1_no_base:.4f}"
        )

def normalize_ambiguity(value) -> str:
    s = "" if pd.isna(value) else str(value)
    s = s.strip().lower()
    return "yes" if s in {"yes", "y", "true", "1"} else "no"

if "Ambiguity" in df.columns:
    df["_amb_norm"] = df["Ambiguity"].apply(normalize_ambiguity)
else:
    df["_amb_norm"] = "no"

summarize_subset(df[df["_amb_norm"] == "yes"], "Ambiguity=yes")
summarize_subset(df[df["_amb_norm"] == "no"], "Ambiguity=no")

# Overall
summarize_subset(df, "Overall")
```
